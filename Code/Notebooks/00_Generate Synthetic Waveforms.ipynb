## Import and Rescale Model

First, we will import the HDF5 (`.h5`) model which contains the following keys:
- `DeltaT`: Temperature Anomaly model $\Delta T$
- `vx`: Flow Component $v_x$
- `vy`: Flow Component $v_y$
- `x`: $x$-axis
- `y`: $y$-axis
- `L`: Typical Lengthscale
- `DeltaT0`: Typical Temperature Anomaly $\Delta T_0$
- `v0`: Typical Flow Speed $v_0$

```python
# Which model do you want to import
Model = "Wiens" # Wiens, Precession, BuoyancyFree
```

```python
# Import Model:
data_path = f"../../Target Models/{Model}.h5"
###############
with h5py.File(str(data_path), "r") as f:
    DeltaT = np.array(f["DeltaT"])
    vx = np.array(f["vx"])
    vy = np.array(f["vy"])
    x = np.array(f["x"])
    y = np.array(f["y"])
    L = f["L"][()]
    DeltaT0 = f["DeltaT0"][()]
    v0 = f["v0"][()]

# typical temperature anomaly
print(f"Typical Temperature Anomaly DT0 = {DeltaT0} K")
# Typical Flow Speed
print(f"Typical Flow Speed v0 = %.1f cm/s" %(v0*100))
# typical lengthscale
print(f"Typical Length Scale L = {L*100} cm")

# Background Temperature
T0 = 300     # [K]

# typical wavelength
lam = L/50 # [m]

# Rescale Model:
x = x/lam
y = y/lam
if v0:
    vx = vx/v0
    vy = vy/v0
if DeltaT0:
    DeltaT = DeltaT/DeltaT0
```

## Physical Model Parameters and Scaling

```python
# Constitutive Relationship
CR = 'IdealGas' #'IdealGas', 'Water'

# Import relevant CR:
sys.path.append(
    '../ConstitutiveRelationships')

if CR == 'IdealGas':
    from CR_IdealGas import c2
    from CR_IdealGas import c_inv
    # typical soundspeed
    c0 = 346     # [m/s]       
else:
    print('wrong constituive relationship')
    
#calculating vp (c) from constituive relationship:
c = np.sqrt(c2(DeltaT, T0, DeltaT0))

# maximum velocity value
c_max = np.max(c) 
    
print('Background Sound Speed: %i m/s' %c0)   # [m/s]
print('Mach Number: %.1e' %(v0/c0))
print('************')
print('Order of Magnitude Phaseshift due to v0 [°]:', 2*L*v0/(c0*lam)*360)
```

## Model Interpolation
Notice that the model is currently fairly coarse (ie. there are relatively few samples in the x- and y-directions).  This discretisation is currently too coarse for the desired source frequency we want to use for this example, so we need to interpolate this coarser model onto a finer grid.

Since the wavelength is fixed to $\lambda = 1$, we need the grid spacing `dh = L/lam/nx` to be desirably smaller than 0.1 - or even smaller since there is higher frequency content in each STF.

A fairly simple way of accomplishing this is to use the `RectBivariateSpline` tool from the `scipy` package.
We can set a desired scaling factor (see `scale`) which will scale the number of samples in each of the x- and y-directions.

```python
scale = 1 # (!) Increase reolution to avoid grid dispersion
# routine to overwrite fields by more high resolved ones

# create spline object - for DeltaT
interp_DT = RectBivariateSpline(np.linspace(0, DeltaT.shape[0], DeltaT.shape[0]), np.linspace(0, DeltaT.shape[1], DeltaT.shape[1]), DeltaT)
# call the spline object on given gridpoints [in pt.-Number] - the larger 'scale', the finer the grid
DeltaT = interp_DT(np.linspace(0, DeltaT.shape[0], DeltaT.shape[0]*scale), np.linspace(0, DeltaT.shape[1], DeltaT.shape[1]*scale))

# repeat for vx
interp_vx = RectBivariateSpline(np.linspace(0, vx.shape[0], vx.shape[0]), np.linspace(0, vx.shape[1], vx.shape[1]), vx)
vx = interp_vx(np.linspace(0, vx.shape[0], vx.shape[0]*scale), np.linspace(0, vx.shape[1], vx.shape[1]*scale))

# repeat for vy
interp_vy = RectBivariateSpline(np.linspace(0, vy.shape[0], vy.shape[0]), np.linspace(0, vy.shape[1], vy.shape[1]), vy)
vy = interp_vy(np.linspace(0, vy.shape[0], vy.shape[0]*scale), np.linspace(0, vy.shape[1], vy.shape[1]*scale))

# rescale plotting grid
x = np.arange(0,DeltaT.shape[0])*(np.max(x)-np.min(x))/np.shape(DeltaT)[0] + np.min(x)
y = np.arange(0,DeltaT.shape[1])*(np.max(y)-np.min(y))/np.shape(DeltaT)[1] + np.min(y)

nx = DeltaT.shape[0]
ny = DeltaT.shape[1]
dh = (np.max(x)-np.min(x))/nx

print(f"New grid step: dh = %.4f" %dh)
print(f"New grid size: Nx = %i" %nx)
print(f"New grid size: Ny = %i" %ny)
```

## Simulation Setup

Next, we can set up the simulation itself.  For this example, we will be explicitly defining the following:
- Time step `dt`
- Number of time steps `nt`
- `ijsrcs`: Indices of the source positions - from which each source is later chosen to emit
- Source time function `sourcetf`
- `ijrecs`: Indices of the receiver positions
- Boundary Conditions and save options

Furthermore, the boolian `SimAqu` defines wether the sources should fire within one simulation or each in its own simulation.

First let us plot the available Transducer Arrays

```python
# Dense Array
Transducers4x4 = SensorArrays.Transducers4x4(nx,ny)
# Sparse Array
Transducers4x2 = SensorArrays.Transducers4x2(nx,ny)
# Irregular Array
Transducers2x2 = SensorArrays.Transducers2x2(nx,ny)

fig = plt.figure(figsize = (12,4))
gs = gridspec.GridSpec(1, 3, width_ratios=[1, 1, 1])

gs.update(wspace = 0.05,
         hspace = 0.2)

ax1 = plt.subplot(gs[0, 0])  # first column, first row
ax2 = plt.subplot(gs[0, 1])  # second column, first row
ax3 = plt.subplot(gs[0, 2])  # third column, first row

SensorArrays.plot(ax1,Transducers4x4,nx,ny,dh)
ax1.set_xlabel("$x/ \\lambda$")
ax1.set_ylabel("$y/ \\lambda$")

SensorArrays.plot(ax2,Transducers4x2,nx,ny,dh)
ax2.set_xlabel("$x/ \\lambda$")
ax2.set_yticks([])

SensorArrays.plot(ax3,Transducers2x2,nx,ny,dh)
ax3.set_xlabel("$x/ \\lambda$")
ax3.set_yticks([])

#plt.savefig("../../Figures/SensorArrays.jpg", dpi = 300, bbox_inches = "tight")
plt.show()
```

```python
### Choose transdcuer Array for Simulation:
Array = "4x4" # 4x4, 4x2, 2x2
###########################################
```

```python
# We assign the source- and receiver indices:
if Array == "4x4":
    ijsrc, ijrec = Transducers4x4
elif Array == "4x2":
    ijsrc, ijrec = Transducers4x2
else:
    ijsrc, ijrec = Transducers2x2
```

```python
# Temporal discretization
dt = 0.008 # (!) larger timestep than in paper to reduce computational expense - lower for higher sensitivity to phase shifts
nt = 35000
t = np.arange(0.0, nt*dt, dt)


# Check grid dispersion and stability 
print('Number of Gridpoints %i x %i' %(nx,ny))
print("Stability criterion, CFL number: %.3f" %(c_max*dt/dh))
###############################################

# Creating the model array
m = np.zeros((3,np.shape(DeltaT)[0],np.shape(DeltaT)[1]))
m[0,:,:] = DeltaT
m[1,:,:] = vx
m[2,:,:] = vy

###############################################
# SOURCES

# Select wether sources should fire Simultaneously (Simultaneous Aquisition):
################
SimAqu = True #
################

# Extracting the source positions
srcpos = ijsrc.T * dh
srcpos[:,0] += np.min(x)
srcpos[:,1] += np.min(y)

# Source time (center)
t0 = 4
# Source center frequency
f0 = 1
# Width of signal (cycles)
wSTF = 3  # Required for sourcetimefuncs.cossource

# Source time function - as an array
sourcetf = np.zeros((ijsrc.shape[1],nt))
envelope = np.zeros((ijsrc.shape[1],nt))

if SimAqu:
    #Perturb Source Time of Shots
    for s in range(ijsrc.shape[1]):
        # delay to t0 may be added to vary source times
        #************
        STF_delay = 0.5*s
        STF_scale = (1.5+0.1*(s-8))
        #************
        sourcetf[s,:], envelope[s,:] = SourceTimeFuncs.cossource(t, t0+STF_delay, f0, wSTF)
        sourcetf[s,:] *= STF_scale
else:
    for s in range(ijsrc.shape[1]):
        sourcetf[s,:], envelope[s,:] = SourceTimeFuncs.cossource(t, t0, f0, wSTF)

# Show STF:
plt.figure(figsize = [5,4])
plt.title('Source Time Function')
plt.plot(t,sourcetf[3,:], label = "$\\phi(t)$ [Pa]", color = "darkblue")
plt.plot(t,envelope[3,:], label = "$E(t)$", color = "lightgreen")
plt.xlim(0,5*t0)
plt.xlabel('$t/\\tau_c$')
plt.legend()
#plt.savefig("../../Figures/STF.jpg", dpi = 300, bbox_inches = "tight")
plt.show()

###############################################
# Extracting receiver coordinates
nrec = ijrec.shape[1]
recpos = ijrec.T * dh
recpos[:,0] += np.min(x)
recpos[:,1] += np.min(y)

###############################################
# Initiallize input parameter dictionary
inpar = {}
inpar["ntimesteps"] = nt
inpar["nx"] = nx
inpar["ny"] = ny
inpar["dt"] = dt
inpar["dh"] = dh
inpar["T0"] = T0            # dimensional for CR_Water
inpar["DeltaT0"] = DeltaT0  # dimensional for CR_Water
inpar["Ma"] = v0/c0
inpar["savesnapshot"] = False
inpar["snapevery"] = 200
inpar["comp_fact"] = 2 # (!) compression factor for saved snapshots (insert divider of both nx and ny) - should be increased for very large grids
inpar["savefinal"] = False
inpar["boundcond"] = "ReflBou"
inpar["constrel"] = CR
```

```python
# Plot the model including source/transducer positions
Plotting.PlotModel(x,y,DeltaT,vx,vy,srcpos = srcpos, v_inc = 12, comp_fact = 4)
```

```python
# Export Sensor positions to apply in a straight ray code:
X_xd = ijsrc*dh*lam

SR_folder = "../../SRI_Export/"
if not os.path.exists(SR_folder):
    os.makedirs(SR_folder)
    print(f"Folder '{SR_folder}' created.")
    
with h5py.File(f'{SR_folder}Sensor_Positions_4x4.h5', 'w') as f:
    f.create_dataset("X_xd", data = X_xd)
```

## Running the Simulation

Now that we have set up the forward simulation, we can run this using `solveacoustic2D` in `pestoseis.USwaveprop2D_on_cuda`.
The Wavforms as well as the snapshots (if `inpar["savesnapshot"]` is true) are stored to `outfileh5`.

```python
# Define the folder where to put the data
folder = f"../../Data/Full-Waveform Data and Reconstructions/{Model} {Array}"

if not os.path.exists(folder):
    os.makedirs(folder)
    print(f"Folder '{folder}' created.")

if SimAqu:
    # Run the simulation with all sources firing
    seism, psave, p_final = WaveProp_on_cuda.solveacoustic2D(inpar, ijsrc, m, sourcetf, ijrec,
                                                             outfileh5 = folder+'/Data_SimAqu_mediumSTF.h5')
else:

    # Run simulations consecutively for specified sources
    N_sensors = ijsrc.shape[1]
    
    seism = np.zeros((N_sensors, N_sensors, nt))
    for s in range(N_sensors):
        print("***")
        print("Forward Simulation for Src No %i" %s)
        print("***")
            
        SRC = ijsrc[:,[s]]
        STF = sourcetf[[s],:]
        
        seism_s, psave_s, p_final_s = WaveProp_on_cuda.solveacoustic2D(inpar, SRC, m, STF, ijrec, saveh5 = False)    
        seism[s,:,:] = seism_s
    
    # Save Results to h5 file:
    outfileh5 = folder + "/Data_SepAqu_mediumSTF.h5"
    hf = h5py.File(outfileh5,"w")
    hf["DeltaT"] = m[0,:,:]
    hf["vx"] = m[1,:,:]
    hf["vy"] = m[2,:,:]
    hf["dh"] = dh
    hf["seism"] = seism
    hf["dt"] = dt
    hf["srcij"] = ijsrc
    hf["recij"] = ijrec
    hf["sourcetf"] = sourcetf
    hf["Ma"] = v0/c0
    hf["T0"] = T0
    hf["DeltaT0"] = DeltaT0
    hf.close()
    
    print("Data saved to " + outfileh5)
```
