# Evaluation of the Inversions
This notebook can be used to load the reconstructions of the FWI and SRI. By doing so, the evolution of the misfit $\chi$ and the choice of the widths $\sigma_k$ of the filters, with which we filter the gradients, can be reviewed.\
Furthermore, this routine is used to average the FWI results onto the straight-ray grid and compute the reconstruction errors.

```python
#import standard packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
# Import specific packages
import h5py
import sys
from matplotlib import gridspec 
from scipy.interpolate import griddata
import scipy.io
import os # for file administration

# Function to recursively print all groups and datasets
def print_structure(name, obj):
    print(name)

# Add the folder containing .py files to the Python path
sys.path.append(
    '../Scripts')

import Plotting
import SensorArrays
```

## Review a Temperature Inversion
Evaluate a **Temperature Reconstruction** of the FWI

```python
Model = "Wiens" # Wiens, BuoyancyFree
Array = "2x2"   # 4x4, 4x2, 2x2

folder = f'../../Data/Full-Waveform Data and Reconstructions/{Model} {Array}'
```

```python
# we construct the folder where to save all the Results:
Results_folder = "../../Data/Results"
if not os.path.exists(Results_folder):
    os.makedirs(Results_folder)
    print(f"Folder '{Results_folder}' created.")
```

```python
####################
# Domain Size
L = 1      # [m]
####################
# sound wavelength
lam = L/50  # [m]
#Sound Speed 
c0 = 346 # [m/s]

# load true model
h5_data = f"{folder}/Data_SimAqu_mediumSTF.h5"

with h5py.File(h5_data,"r") as h:
    DeltaT0 = h["DeltaT0"][()]
    dh = h["dh"][()] 
    DeltaT = h["DeltaT"][:,:].copy()
    
# Make dimensional
DeltaT *= DeltaT0
dh *= lam
```

We want to view the Evolution of the misfit $\chi_k$ and the kernel widths $\sigma_k$ of the filters at each iteration $k$ up to the last performed `Iteration`.

```python
Iteration = 18

# load Information on FWI
h5_data = folder+f"/Iteration-{Iteration}_SimAqu.h5"
with h5py.File(h5_data,"r") as h:
    #----------------------------
    Chi = h["Misfit_history"][:].copy()
    for_Temperature = h["for_Temperature"][:].copy()
    filtT = h["T_filter"][:].copy()


plt.rcParams.update({'font.size': 12})

fig, ax1 = plt.subplots(figsize=(6, 3.3))

# Plot the first quantity with a logarithmic y-axis
color = "orangered"
ax1.set_yscale('log')
ax1.plot(Chi, color = color, linestyle = " ", marker = "x", markersize = 11, label='Exponential')
#ax1.set_xlabel('Iteration Index $k$')
ax1.set_ylabel("$\\chi(m_k)$", color = color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xlabel("Iteration $k$")

color = "midnightblue"
ax2 = ax1.twinx()
ax2.plot(filtT[:], color = color, linestyle = " ", marker = "+", markersize = 12)
ax2.set_ylabel('$\\sigma_\\text{k}/ \\lambda$', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yticks([1,2,3,4,5])
ax2.set_xticks(range(Iteration+1))

fig.tight_layout()
plt.show()
```

Since the **misfit history and the information on the filter widths** is stored alongside the last few (large-sized) gradients, we extract the most essential information to a small file:

```python
###################################################################
# save this information to the Results Folder?
StoreInfo = False
# Create a new file or append an existing file?
CreateNewFile = False
###################################################################
fname = f"{Results_folder}/{Model}_FWI_Information.h5"
if StoreInfo:
        
    # switch from "a" (append) to "w" write to (re-)create a file
    if CreateNewFile:
        method = "w" 
    else: 
        method = "a" 

    # save information
    with h5py.File(fname, method) as hf:
        hf[f"for_Temperature_{Array}"] = for_Temperature
        hf[f"sigma_DTfilter_{Array}"] = filtT*lam
        if f"Misfit_hist_{Array}" not in hf.keys():
            hf[f"Misfit_hist_{Array}"] = Chi

f = h5py.File(fname, "r")
#Let us visit 
print("Keys in Wrapped file: \n ", [key for key in f.keys()])
f.close()
```

We can now load the estimated model created by the routine `Full-Waveform Inversion.ipynb` as well as the straight ray reconstruction. \
Then we **average the true model and the FWI reconstruction onto the grid of the SRI, compute the error and...**

```python
# load estimated model
h5_data = f"{folder}/Data_m{Iteration}_SimAqu.h5"

with h5py.File(h5_data,"r") as h:
    DeltaT_FWI = h["DeltaT"][:,:].copy()

DeltaT_FWI *= DeltaT0
```

```python
# load Straight Ray model
folder_SR = "../../Data/Straight-Ray Reconstructions"

if Model == "BuoyancyFree":
    SR_data = scipy.io.loadmat(f"{folder_SR}/SRmodel_{Model}_{Array}_DT.mat")
else:
    SR_data = scipy.io.loadmat(f"{folder_SR}/SRmodel_{Model}_{Array}.mat")

nx_SR = int(SR_data["N_plot2"][0,0])
ny_SR = int(SR_data["N_plot2"][0,1])
DeltaT_SR = SR_data["T_hat2"].reshape(nx_SR,ny_SR)

# Compute the relation between FWI-grid and SRI-gird
ratio_x = DeltaT.shape[0]//nx_SR
ratio_y = DeltaT.shape[1]//ny_SR

x_SR = np.arange(0,nx_SR)*dh*ratio_x
y_SR = np.arange(0,ny_SR)*dh*ratio_y

# Average True Model
DeltaT_avg = DeltaT.reshape(nx_SR, ratio_x, ny_SR, ratio_y).mean(axis = (1,3))
# ... and FWI reconstruction
DeltaT_FWI_avg = DeltaT_FWI.reshape(nx_SR, ratio_x, ny_SR, ratio_y).mean(axis = (1,3))
```

```python
# Calculate both RRSEs
N_pts = nx_SR*ny_SR
DT_FWI_RRSE = np.sqrt(np.sum((DeltaT_avg-DeltaT_FWI_avg)**2)/np.sum((DeltaT_avg - DeltaT_avg.mean())**2.))
DT_SR_RRSE = np.sqrt(np.sum((DeltaT_avg-DeltaT_SR)**2)/np.sum((DeltaT_avg -DeltaT_avg.mean())**2.))

# Print the reconstruction errors:
print("RRSE of the FWI: %.2f" %(DT_FWI_RRSE))
print("RRSE of the SRI: %.2f" %(DT_SR_RRSE))
```

**... save the results** into a .h5-file

```python
# Location
fname = f"{Results_folder}/{Model}_Wrapped.h5"

###################################################################
# Store Results? And create a new file or append an existing file?
StoreResults = False
CreateNewFile = False
###################################################################

if StoreResults:
    # switch from "a" (append) to "w" write to (re-)create a file
    if CreateNewFile:
        method = "w" 
    else: 
        method = "a" 

    # save information
    with h5py.File(fname, method) as hf:  
        if "DeltaT_true" not in hf.keys():
            hf["DeltaT_true"] = DeltaT_avg
        if "x" not in hf.keys():
            hf["x"] = x_SR
            hf["y"] = y_SR
        hf[f"DeltaT_{Array}_FWI"] = DeltaT_FWI_avg
        hf[f"DeltaT_{Array}_SRI"] = DeltaT_SR
    
        hf[f"DeltaT_Error_{Array}_FWI"] = DT_FWI_RRSE
        hf[f"DeltaT_Error_{Array}_SRI"] = DT_SR_RRSE

f = h5py.File(fname, "r")
#Let us visit 
print("Keys in Wrapped file: \n ", [key for key in f.keys()])
f.close()
```

Now we **compare the FWI Reconstruction** (averaged onto $128 \times 128$ grid points) **to the corresponding Straight-Ray Reconstruction** \
Both are stored in the File inside the 'Results' folder:

```python
# Read File
fname = f"{Results_folder}/{Model}_Wrapped.h5"

Methods = ["FWI", "SRI"] 
Reconstructions = []
Errors = []

with h5py.File(fname, "r") as hf:
    x = hf["x"][:]
    y = hf["y"][:]
    
    DeltaT = hf["DeltaT_true"][:,:]
    
    for Method in Methods:
        Reconstructions.append(hf[f"DeltaT_{Array}_{Method}"][:,:])
        Errors.append(hf[f"DeltaT_Error_{Array}_{Method}"])

if Array == "4x4": 
    ijsrc, ijrec = SensorArrays.Transducers4x4(len(x),len(y))
elif Array == "4x2":
    ijsrc, ijrec = SensorArrays.Transducers4x2(len(x),len(y))
elif Array == "2x2":
    ijsrc, ijrec = SensorArrays.Transducers2x2(len(x),len(y))

srcpos = ijsrc.T*(x[1]-x[0])
```

## Review a Velocity Inversion
Evaluate a **Velocity Reconstruction** of the FWI.\
We will repeat the procedure of the code above.

```python
Model = "Precession" # Precession, BuoyancyFree
Array = "2x2"

folder = f'../../Data/Full-Waveform Data and Reconstructions/{Model} {Array}'
```

```python
####################
# Domain Size
L = 1     # [m]
####################
# sound wavelength
lam = L/50  # [m]
#Sound Speed 
c0 = 346 # [m/s]

# load true model
h5_data = f"{folder}/Data_SimAqu_mediumSTF.h5"

with h5py.File(h5_data,"r") as h:
    vx = h["vx"][:,:].copy()
    vy = h["vy"][:,:].copy()
    dh = h["dh"][()]
    Ma = h["Ma"][()]

# Make Dimensional
v0 = Ma*c0
vx *= v0
vy *= v0
dh *= lam
```

Evolution of the misfit $\chi_k$ and the kernel widths $\sigma_k$ of the filters at each iteration $k$ up to the last performed `Iteration`.

```python
###############
Iteration = 12
###############

# load Information on FWI
h5_data = folder+f"/Iteration-{Iteration}_SimAqu.h5"
with h5py.File(h5_data,"r") as h:
    #----------------------------
    Chi = h["Misfit_history"][:].copy()
    for_Velocity = h["for_Velocity"][:].copy()
    filtV = h["v_filter"][:].copy()


plt.rcParams.update({'font.size': 12})
fig, ax1 = plt.subplots(figsize=(6, 3.3))

# Plot the first quantity with a logarithmic y-axis
color = "limegreen"
ax1.set_yscale('log')
ax1.plot(Chi, color = color, linestyle = " ", marker = "x", markersize = 11, label='Exponential')
ax1.set_xlabel('Iteration Index $k$')
ax1.set_ylabel("$\\chi(m_k)$", color = color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xlabel("Iteration $k$")

color = "darkblue"
ax2 = ax1.twinx()
ax2.plot(filtV[:], color = color, linestyle = " ", marker = "+", markersize = 12)
ax2.set_ylabel('$\\sigma_\\text{k}$', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_yticks([1,2,3,4,5])
ax2.set_xticks(range(Iteration+1))
plt.show()
```

```python
# load estimated model
h5_data = f"{folder}/Data_m{Iteration}_SimAqu.h5"

with h5py.File(h5_data,"r") as h:
    vx_FWI = h["vx"][:,:].copy()
    vy_FWI = h["vy"][:,:].copy()

vx_FWI *= v0
vy_FWI *= v0
```

```python
# load Straight Ray model
folder_SR = "../../Data/Straight-Ray Reconstructions"

if Model == "BuoyancyFree":
    SR_data = scipy.io.loadmat(f"{folder_SR}/SRmodel_{Model}_{Array}_v.mat")
else:
    SR_data = scipy.io.loadmat(f"{folder_SR}/SRmodel_{Model}_{Array}.mat")

nx_SR = int(SR_data["N_plot2"][0,0])
ny_SR = int(SR_data["N_plot2"][0,1])
vx_SR = SR_data["u_hat2"].reshape(nx_SR,ny_SR)
vy_SR = SR_data["v_hat2"].reshape(nx_SR,ny_SR)

# Establish grid for Straight-Ray Model
ratio = vx.shape[0]//nx_SR
x_SR = np.arange(0,nx_SR)*dh*ratio
y_SR = np.arange(0,ny_SR)*dh*ratio

# Average True Model...
vx_avg = vx.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))
vy_avg = vy.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))

# ... and FWI reconstruction
vx_FWI_avg = vx_FWI.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))
vy_FWI_avg = vy_FWI.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))

# Calculate Magnitude of v
v_avg = np.sqrt(vx_avg**2 + vy_avg**2)
v_FWI_avg = np.sqrt(vx_FWI_avg**2 + vy_FWI_avg**2)
v_SR = np.sqrt(vx_SR**2 + vy_SR**2)
```

```python
N_pts = (nx_SR*ny_SR)

# RMS error of FWI_model
v_FWI_RRSE = np.sqrt(np.sum((vx_avg-vx_FWI_avg)**2 + (vy_avg-vy_FWI_avg)**2)
                           /np.sum((vx_avg - vx_avg.mean())**2. + (vy_avg - vy_avg.mean())**2.))
# RMS error of SRI_model
v_SR_RRSE = np.sqrt(np.sum((vx_avg-vx_SR)**2 + (vy_avg-vy_SR)**2)
                          /np.sum((vx_avg - vx_avg.mean())**2. + (vy_avg - vy_avg.mean())**2.))

# Print the reconstruction errors:
print("RRSE of the FWI: %.2f" %(v_FWI_RRSE))
print("RRSE of the SRI: %.2f" %(v_SR_RRSE))
```

We again **save the wrapped-up results** in a .h5-file:

```python
# location
fname = f"{Results_folder}/{Model}_Wrapped.h5"

###################################################################
# Store Results? And create a new file or append an existing file?
StoreResults = False
CreateNewFile = False
###################################################################

if StoreResults:
    # switch from "a" (append) to "w" write to (re-)create a file
    if CreateNewFile:
        method = "w" 
    else: 
        method = "a"

    with h5py.File(fname, method) as hf:
        if "vx_true" not in hf.keys():
            hf["vx_true"] = vx_avg
            hf["vy_true"] = vy_avg
        if "x" not in hf.keys():
            hf["x"] = x_SR
            hf["y"] = y_SR
        hf[f"vx_{Array}_FWI"] = vx_FWI_avg
        hf[f"vy_{Array}_FWI"] = vy_FWI_avg
        hf[f"vx_{Array}_SRI"] = vx_SR
        hf[f"vy_{Array}_SRI"] = vy_SR
    
        hf[f"v_Error_{Array}_FWI"] = v_FWI_RRSE
        hf[f"v_Error_{Array}_SRI"] = v_SR_RRSE

f = h5py.File(fname, "r")
print("Keys in Wrapped file: \n ", [key for key in f.keys()])
f.close()
```

Now we **compare the FWI Reconstruction** (averaged onto $128 \times 128$ grid points) **to the corresponding Straight-Ray Reconstruction** \
Both are stored in the File inside the 'Results' folder:

```python
# Read File
fname = f"{Results_folder}/{Model}_Wrapped.h5"

Methods = ["FWI", "SRI"]
Fields = ["vx", "vy"]
Reconstructions = []
Errors = []

with h5py.File(fname, "r") as hf:
    x = hf["x"][:]
    y = hf["y"][:]
    
    vx = hf["vx_true"][:,:]
    vy = hf["vy_true"][:,:]    
    for Method in Methods:
        Errors.append(hf[f"v_Error_{Array}_{Method}"])
        for Field in Fields:
            Reconstructions.append(hf[f"{Field}_{Array}_{Method}"][:,:])

if Array == "4x4": 
    ijsrc, ijrec = SensorArrays.Transducers4x4(len(x),len(y))
elif Array == "4x2":
    ijsrc, ijrec = SensorArrays.Transducers4x2(len(x),len(y))
elif Array == "2x2":
    ijsrc, ijrec = SensorArrays.Transducers2x2(len(x),len(y))

srcpos = ijsrc.T*(x[1]-x[0])
```
