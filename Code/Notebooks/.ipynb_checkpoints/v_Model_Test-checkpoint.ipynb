```python
import numpy as np
import matplotlib.pyplot as plt
import sys

# package for efficient array storage
import h5py
```

```python
def Test(x,y):
    
    v = np.zeros((2, len(x), len(y))) #2D Vector field (flow velocity)
    for i in range(len(x)):
        for j in range(len(y)):
            v[0,i,j] = 0.
            v[1,i,j] = 1.
    
    return v
```

## Cyclonic Windfield: $v = r^2 e_\varphi$

```python
# grid point numbers
nx = 150
ny = 100

x = np.linspace(-nx/2,nx/2,nx)
y = np.linspace(-ny/2,ny/2,ny)

# temperature distribution
v = Test(x,y)/np.max(Test(x,y))
```

```python
# Select every ... grid point for v representation
incr = 3 # increment: every ... v-value

vx_every = v[0, ::incr, ::incr]
vy_every = v[1, ::incr, ::incr]

# Create grid for plotting
x_every = x[::incr]
y_every = y[::incr]

plt.quiver(x_every,y_every,vx_every.T, vy_every.T)
plt.xlabel('x')
plt.ylabel('y')
```

# Uniform zero Temperature dist.

```python
DeltaT = np.zeros((len(x),len(y)))
```

### Saving the model as .h5-file

```python
# Create an HDF5 file in write mode ('w')
with h5py.File('v_model_Test.h5', 'w') as f:
    f.create_dataset('DeltaT', data = DeltaT)
    f.create_dataset('vx', data = v[0,:,:])
    f.create_dataset('vy', data = v[1,:,:])
```
