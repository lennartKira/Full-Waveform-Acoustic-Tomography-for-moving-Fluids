### Plot the Graphical Abstract

```python
#import standard packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
# Import specific packages
import pandas as pd
import h5py
import sys
from matplotlib import gridspec 
from scipy.interpolate import griddata
import scipy.io
import os # for file administration
```

```python
# Define background Soundspeed
c0 = 346
#Define Wavelength:
lam = 0.02

# load velocity data of very sparse array:
folder = '../../Data/Full-Waveform Data and Reconstructions/Precession 2x2'

# load true model
h5_data = folder+f"/Data_SimAqu_mediumSTF_Tr.h5"
print('True Data:', h5_data)
h=h5py.File(h5_data,"r")
Ma = h["Ma"][()]
vx = h["vx"][:,:].copy()*Ma*c0
vy = h["vy"][:,:].copy()*Ma*c0
dt = h["dt"][()]*lam/c0
dh = h["dh"][()]*lam
h.close()

# load FWI model
h5_data = folder+f"/Data_m12_SimAqu_Tr.h5"
h=h5py.File(h5_data,"r")
vx_FWI = h["vx"][:,:].copy()*Ma*c0
vy_FWI = h["vy"][:,:].copy()*Ma*c0
srcpos = h["srcij"][:,:].copy().T*dh
h.close()
```

```python
# load Straight Ray model
folder_SR = "../../Data/Straight-Ray Analysis"
SR_data = scipy.io.loadmat(f"{folder_SR}/SRmodel_Precession_2x2_s9.mat")

nx_SR = int(SR_data["N_plot2"][0,0])
ny_SR = int(SR_data["N_plot2"][0,1])
vx_SR = SR_data["u_hat2"].reshape(nx_SR,ny_SR)
vy_SR = SR_data["v_hat2"].reshape(nx_SR,ny_SR)
RBF_centers = SR_data["Xc_i"]

# Establish grid for Straight-Ray Model
ratio = vx.shape[0]//nx_SR
x_SR = np.arange(0,nx_SR)*dh*ratio
y_SR = np.arange(0,ny_SR)*dh*ratio

# Average True Model...
vx_avg = vx.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))
vy_avg = vy.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))
# ... and FWI reconstruction
vx_FWI_avg = vx_FWI.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))
vy_FWI_avg = vy_FWI.reshape(nx_SR, ratio, ny_SR, ratio).mean(axis = (1,3))

# Calculate Magnitude of v
v_avg = np.sqrt(vx_avg**2 + vy_avg**2)
v_FWI_avg = np.sqrt(vx_FWI_avg**2 + vy_FWI_avg**2)
v_SR = np.sqrt(vx_SR**2 + vy_SR**2)
```

```python
# Plot the Graphical Abstract
v_inc = 8 # velocity quiver increment

fig = plt.figure(figsize = (11.3,3.5))
gs = gridspec.GridSpec(1, 4, width_ratios=[1, 1, 1, 0.06])

gs.update(wspace = 0.1,
         hspace = 0.2)

ax1 = plt.subplot(gs[0, 0])  # first column, first row
ax2 = plt.subplot(gs[0, 1])  # second column, first row
ax3 = plt.subplot(gs[0, 2])  # second column, first row

clim = [np.min([v_avg]), np.max([v_avg])] # v_FWI, v_SR

im1 = ax1.pcolormesh(x_SR,y_SR,v_avg.T, cmap="viridis", clim = clim)
ax1.quiver(x_SR[::v_inc],y_SR[::v_inc],vx_avg[::v_inc,::v_inc].T,vy_avg[::v_inc,::v_inc].T, color = "snow", width = 4.5e-3, scale = 0.55)
ax1.set_xlabel('$x$ [m]')
ax1.set_ylabel("$y$ [m]")
ax1.set_title('True Flow Field and Direct Rays')
ax1.set_aspect('equal')

im2 = ax2.pcolormesh(x_SR,y_SR,v_FWI_avg.T, cmap="viridis", clim = clim)
ax2.quiver(x_SR[::v_inc],y_SR[::v_inc],vx_FWI_avg[::v_inc,::v_inc].T,vy_FWI_avg[::v_inc,::v_inc].T, color = "snow", width = 4.5e-3, scale = 0.55)
ax2.set_xlabel('$x$ [m]')
ax2.set_yticks([])
ax2.set_title('Full-Waveform Inversion')
ax2.set_aspect('equal')

im3 = ax3.pcolormesh(x_SR,y_SR,v_SR[:,:].T, cmap="viridis", clim = clim)
ax3.quiver(x_SR[::v_inc],y_SR[::v_inc],vx_SR[::v_inc,::v_inc].T,vy_SR[::v_inc,::v_inc].T, color = "snow", width = 4.5e-3, scale = 0.55)
ax3.set_xlabel('$x$ [m]')
ax3.set_yticks([])
ax3.set_title('Straight-Ray Inversion')
ax3.set_aspect('equal')
plt.gca().set_aspect('equal')
#if srcpos is not None:
#    ax1.scatter(srcpos[:,0], srcpos[:,1], c = "lightcyan", marker = "x", s = 100, linewidths = 4)
#    ax1.set_xlim(x.min(), x.max())
#    ax1.set_ylim(y.min(), y.max())
#    ax2.scatter(srcpos[:,0], srcpos[:,1], c = "lightcyan", marker = "x", s = 100, linewidths = 4)
#    ax2.set_xlim(x.min(), x.max())
#    ax2.set_ylim(y.min(), y.max()) 

# Colorbar
cbar_ax = plt.subplot(gs[0, 3])  # Dedicated axis for the colorbar
cbar = fig.colorbar(im2, cax=cbar_ax, label='$v$ [m/s]')

# Adding the array
sources = srcpos.T
receivers = srcpos.T
for i in range(sources.shape[1]):
    for j in range(receivers.shape[1]):
        ax1.plot([sources[0,i], receivers[0,j]], [sources[1,i], receivers[1,j]], color = "azure", linestyle = "--", zorder = 1)
ax1.scatter(sources[0, :], sources[1,:], color = "white", marker = "*", s = 140, zorder = 2)
ax1.set_xlim((0,x_SR.max()))
ax1.set_ylim((0,y_SR.max()))
ax1.set_aspect('equal')

plt.savefig("../../Figures/GraphicalAbstract.jpg", dpi = 300, bbox_inches='tight')
plt.show()
```
