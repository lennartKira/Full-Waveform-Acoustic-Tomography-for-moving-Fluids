# Compute Sensitivity Kernels and Gradients as an Example
Using the true data and a prior model, this routine computes the Gradients of the Misfit and displays them for demonstrative purposes.\
The true data loaded here should be synthetic data generated by using a specific source-receiver pair.

```python
#import standard packages
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.size"] = 11
```

```python
# Import specific packages
import h5py
import sys
from matplotlib import gridspec
import os # for file administration
import time

# Add the folder containing .py files to the Python path
sys.path.append(
    '../Scripts')

# Import Custom Modules
import WaveProp_on_cuda
import InverseUtils
import AnimateGradients
import Plotting
```

### Material Equation

```python
CR = "IdealGas"

# Import relevant CR:
sys.path.append(
    '../ConstitutiveRelationships')

if CR == 'IdealGas':
    from CR_IdealGas import c2
    from CR_IdealGas import c_inv
    # typical soundspeed
    c0 = 346     # [m/s]       
else:
    print('wrong constituive relationship')
```

### Import True Waveform Data 
Read the data produced from the foward simulation

```python
# Choose folder
folder = '../../Data/Full-Waveform Data and Reconstructions/Precession'
```

```python
# Choose source and receiver
s = [2]
r = [7]

# load true data and compute misfit
h5_data = folder+f"/Data_src-{s}_rec-{r}_mediumSTF.h5"
print('True Data:', h5_data)
with h5py.File(h5_data,"r") as h:
    # import true waveforms
    #----------------------
    seism0 = h["seism"][:,:].copy()
    dt = h["dt"][()]
    
    #import true model
    DeltaT = h["DeltaT"][:,:].copy()
    vx = h["vx"][:,:].copy()
    vy = h["vy"][:,:].copy()
    dh = h["dh"][()]
    srcpos = h["srcij"][:,:].T*dh
    recpos = h["recij"][:,:].T*dh

m = np.zeros((3,DeltaT.shape[0],DeltaT.shape[1]))

m[0,:,:] = DeltaT
m[1,:,:] = vx
m[2,:,:] = vy
```

### Input Scaling and Sourcetimefunction
For now, read from synthetic data as well

```python
# Read from a Simulation
print('True Data:', h5_data)
h=h5py.File(h5_data,"r")

# import true waveforms
#----------------------
Ma = h["Ma"][()]
T0 = h["T0"][()]
DeltaT0 = h["DeltaT0"][()]

sourcetf = h["sourcetf"][:,:].copy()
h.close()

N_sensors, nt = sourcetf.shape

t = np.arange(nt)*dt

# sound wavelength
lam = 4e-2
```

### Define Prior Model and Simulation Parameters
For now, read geometry from true model

```python
# load model geometry
print('Model Geometry from:', h5_data)
h=h5py.File(h5_data,"r")

nx, ny = h["DeltaT"][:,:].copy().shape
dh = h["dh"][()]
ijsrc = h["srcij"][:,:].copy()
ijrec = h["recij"][:,:].copy()
h.close()

# zero prior model
m_prior = np.zeros((3,nx,ny))

# rescale plotting grid
x = np.arange(0,nx)*dh
y = np.arange(0,ny)*dh


# Initialize input parameter dictionary for prior simulations
inpar = {}
inpar["ntimesteps"] = nt
inpar["nx"] = nx
inpar["ny"] = ny
inpar["dt"] = dt
inpar["dh"] = dh
inpar["T0"] = T0            # dimensional for CR_Water
inpar["DeltaT0"] = DeltaT0  # dimensional for CR_Water
inpar["Ma"] = Ma
inpar["savesnapshot"] = False
inpar["snapevery"] = 1000
inpar["savefinal"] = True
inpar["comp_fact"] = 4
inpar["boundcond"] = "ReflBou"
inpar["constrel"] = CR

x.shape
```

## Run Prior Simulation or load Prior Waveform Data

```python
priorfile_ab = folder + f"/Data_m0_src-{s}_rec-{r}_mediumSTF.h5"

# Generate Prior Waveforms?
Run_Prior_Simulation = False
#********************************************

if Run_Prior_Simulation:
    seism, p_save, p_final = WaveProp_on_cuda.solveacoustic2D(inpar, ijsrc, m_prior, sourcetf, ijrec,
                                                              saveh5 = True, outfileh5 = priorfile_ab)
    
else:
    print('Prior Model Data from:', priorfile_ab)
    h=h5py.File(priorfile_ab,"r")

    #----------------------------
    seism = h["seism"][:,:].copy()
    p_final = h["p_final"][:,:,:].copy()
    h.close()
```

## Compute and Display Adjoint Source
From the true and prior waveforms the adjoint source can be generated and displayed

```python
adj_src = InverseUtils.adjointsource_L2(seism0, seism)

#%matplotlib widget

f = plt.figure(figsize = (12,4))
gs = gridspec.GridSpec(1, 10)

gs.update(wspace = 0.4, hspace = 0.3)

ax1 = plt.subplot(gs[0, 0:5])  # first column
ax2 = plt.subplot(gs[0, 6:10])  # second column, first row

ax1.plot(t, seism[0,:], color = "orangered", label = "prior data")
ax1.plot(t, seism0[0,:], color = "darkblue", label = "true data")
#ax1.set_xlim(63,63.5)
ax1.set_xlabel("$t$")
ax1.set_ylabel("$p$", rotation = 0)
ax1.legend(fontsize = 11)

ax2.plot(t, adj_src[0,:], color = "k")
#ax2.set_xlim(55,68)
ax2.set_xlabel("$t$")
ax2.set_ylabel("$\\frac{\partial \chi_1}{\partial p}$", rotation = 0, fontsize = 18)
```

## Compute, Filter and Animate the Gradient
Including animation on how the sensitivity kernel is formed.

```python
gradfile_ab = folder+f"/Gradient_src-{s}_rec-{r}.h5"

# Compute Gradient here?
Run_Inverse_Simulation = False

if Run_Inverse_Simulation:
    inpar["savesnapshot"] = True
    inpar["snapevery"] = 500
    grad_ab = InverseUtils.computegrad2D(inpar, ijrec, m_prior, adj_src,
                                         p_final, saveh5=True, outfileh5 = gradfile_ab)
else:
    # load file with pressure and adjoint field
    h=h5py.File(gradfile_ab,"r")
    grad_ab = np.zeros((3,nx,ny))
    grad_ab[0,:,:] = h["gradient_c"]
    grad_ab[1,:,:] = h["gradient_vx"]
    grad_ab[2,:,:] = h["gradient_vy"]
    h.close()
```

Display the Gradient (without smoothing)

```python
Plotting.ModelvGradient(x,y,m,grad_ab,"v_x",srcpos=srcpos, recpos=recpos, comp_fact = 1, v_inc = 100)
```

Smooth the Gradient first and then Display

```python
# Apply Gaussian Smoothing kernel with standard deviation 'filtT/V' [wavelengths]
# For DeltaT
filtT = 1
# For v
filtV = 1

grad_ab_smooth = np.zeros(grad_ab.shape)

# Smooth first gradients
grad_ab_smooth[0,:,:] = InverseUtils.filter_grad_zeropadding(grad_ab[0,:,:], dh, lambda_min = filtT, pad_width = int(1/dh))
grad_ab_smooth[1,:,:] = InverseUtils.filter_grad_zeropadding(grad_ab[1,:,:], dh, lambda_min = filtV, pad_width = int(1/dh))
grad_ab_smooth[2,:,:] = InverseUtils.filter_grad_zeropadding(grad_ab[2,:,:], dh, lambda_min = filtV, pad_width = int(1/dh))

Plotting.ModelvGradient(x,y,m,grad_ab_smooth,"v_x",srcpos=srcpos, recpos=recpos,comp_fact = 1, v_inc = 100)
```

## Compute the Gradient for the reciprocal shot.
For this, quickly load the true data, compute the prior waveforms and get the adjoint source.

```python
# Load true data and source-receiver positions
h5_data = folder+"/Data_src-[7]_rec-[2]_mediumSTF.h5"
print('True Data:', h5_data)
h=h5py.File(h5_data,"r")
# import true waveforms
seism0 = h["seism"][:,:].copy()
dt = h["dt"][()]
#import source-receiver info
ijsrc = h["srcij"][:,:].copy()
ijrec = h["recij"][:,:].copy()
sourcetf = h["sourcetf"][:,:].copy()
h.close()

inpar["savesnapshot"] = False

priorfile_ba = folder + "/Data_m0_src-[7]_rec-[2]_mediumSTF.h5"
# Generate Prior Waveforms?
Run_Prior_Simulation = False

if Run_Prior_Simulation:
    seism, p_save, p_final = WaveProp_on_cuda.solveacoustic2D(inpar, ijsrc, m_prior, sourcetf, ijrec,
                                                              saveh5 = True, outfileh5 = priorfile_ba)
else:
    # read prior waveforms
    print('Prior Model Data from:', priorfile_ba)
    h=h5py.File(priorfile_ba,"r")

    #----------------------------
    seism = h["seism"][:,:].copy()
    p_final = h["p_final"][:,:,:].copy()
    h.close()

# compute Adjoint Source
adj_src = InverseUtils.adjointsource_L2(seism0, seism)

gradfile_ba = folder+"/Gradient_src-3_rec-0.h5"
# Compute Gradient now?
Run_Inverse_Simulation = False

if Run_Inverse_Simulation:
    grad_ba = InverseUtils.computegrad2D(inpar, ijrec, m_prior, adj_src,
                                         p_final, saveh5=True, outfileh5 = gradfile_ba)
else:
    # load file with pressure and adjoint field
    h=h5py.File(gradfile_ba,"r")
    grad_ba = np.zeros((3,nx,ny))
    grad_ba[0,:,:] = h["gradient_c"]
    grad_ba[1,:,:] = h["gradient_vx"]
    grad_ba[2,:,:] = h["gradient_vy"]
    h.close()
```

```python
grad_ba_smooth = np.zeros(grad_ab.shape)

# Smooth first gradients
grad_ba_smooth[0,:,:] = InverseUtils.filter_grad(grad_ba[0,:,:], dh, lambda_min = filtT)
grad_ba_smooth[1,:,:] = InverseUtils.filter_grad(grad_ba[1,:,:], dh, lambda_min = filtV)
grad_ba_smooth[2,:,:] = InverseUtils.filter_grad(grad_ba[2,:,:], dh, lambda_min = filtV)


#Sum of Gradients
grad_smooth = grad_ab + grad_ba

# Cut out sensors in gradient?
cut_sensors = True
sigma_c = 5/dh

if cut_sensors:
    grad_smooth[0,:,:] = (grad_ab[0,:,:] + grad_ba[0,:,:])*InverseUtils.sens_cutmask(nx, ny, np.hstack((ijrec, ijsrc_pool)), sigma_c)
    grad_smooth[1,:,:] = (grad_ab[1,:,:] + grad_ba[1,:,:])*InverseUtils.sens_cutmask(nx, ny, np.hstack((ijrec, ijsrc_pool)), sigma_c)
    grad_smooth[2,:,:] = (grad_ab[2,:,:] + grad_ba[2,:,:])*InverseUtils.sens_cutmask(nx, ny, np.hstack((ijrec, ijsrc_pool)), sigma_c)

# Smooth gradients
grad_smooth[0,:,:] = InverseUtils.filter_grad(grad_smooth[0,:,:], dh, lambda_min = filtT)
grad_smooth[1,:,:] = InverseUtils.filter_grad(grad_smooth[1,:,:], dh, lambda_min = filtV)
grad_smooth[2,:,:] = InverseUtils.filter_grad(grad_smooth[2,:,:], dh, lambda_min = filtV)

# component of gradient:
model_param = 0 # 0=c, 1=v_x, 2=v_y

# display gradient:
Plotting.PlotReciprocal_Grad(x,y,grad_ab_smooth,grad_ba_smooth,grad_smooth,model_param)
```

```python
# For DeltaT
sigma = 1

gradfile = f"{folder}/Gradient_m0_SimAqu.h5"
# load file with gradient
with h5py.File(gradfile, "r") as h:
    nx, ny = h["gradient_c"].shape
    grad = np.zeros((3,nx,ny))
    grad[0,:,:] = h["gradient_c"]
    grad[1,:,:] = h["gradient_vx"]
    grad[2,:,:] = h["gradient_vy"]
    dh = h["dh"][()]

grad_smooth = np.zeros(grad.shape)
grad_smooth[0,:,:] = InverseUtils.filter_grad(grad[0,:,:], dh, lambda_min = sigma)
grad_smooth[1,:,:] = InverseUtils.filter_grad(grad[1,:,:], dh, lambda_min = sigma)
grad_smooth[2,:,:] = InverseUtils.filter_grad(grad[2,:,:], dh, lambda_min = sigma)
```
