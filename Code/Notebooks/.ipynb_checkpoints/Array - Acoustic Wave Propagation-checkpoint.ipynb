## Typical model values:

```python
# typical temperature anomaly
DeltaT0 = 10 # [K]
# Background Temperature
T0 = 300     # [K]
# Typical Flow Speed
v0 = 1e-1  # [m/s]
# typical lengthscale
L = 0.5      # [m]
# sound wavelength
lam = 1e-2

# Constituive Relationship
CR = 'IdealGas' #'IdealGas' - 'Water' still in progress

# Choose Density effects to be included
Density_effects = False # Set to True for density effects

# Import relevant CR:
sys.path.append(
    '../ConstituiveRelationships')

if CR == 'IdealGas':
    from CR_IdealGas import c2
    # expansion coeff
    alpha = 1/T0 # [1/K] might be non-const. for DeltaT0/T0->1
    # typical soundspeed
    c0 = 300     # [m/s]
        
elif CR == 'Water':
    from CR_Water import c2
    from CR_Water import c2_dim
    # expansion coefficient
    alpha = 2e-4 # [1/K] Putintsev 2005
    # typical soundspeed
    c0 = np.sqrt(c2_dim(DeltaT, T0, DeltaT0)) # [m/s]        
else:
    print('wrong constituive relationship')
    
print('Vol. Expansion Coeff. %.1e 1/K' %alpha)

if not Density_effects:
    alpha = 0
    
    
print('Background Sound Speed: %i m/s' %c0)   # [m/s]
print('Mach Number: %.1e' %(v0/c0))
print('************')
print('Order of Magnitude Phaseshift [°]:', 2*L*v0/(c0*lam)*360)
```

## Model Import

First, we will import the HDF5 (`.h5`) model which contains the following keys:
- `DeltaT`: Temperature Anomaly model $\Delta T$
- `vx`: Flow Component $v_x$
- `vy`: Flow Component $v_y$
- `vp`: P-wave velocity model (c)

```python
data_path = "Cyclone_model.h5"
with h5py.File(str(data_path), "r") as f:
    DeltaT = np.array(f["DeltaT"])
    vx = np.array(f["vx"])
    vy = np.array(f["vy"])
    
#calculating vp (c) from constituive relationship:
vp = np.sqrt(c2(DeltaT, T0, DeltaT0))

# maximum velocity value
vp_max = np.max(vp)
print(vp_max)
```

We can plot the velocity and flow model to verify that the data has been imported correctly.

```python
x = np.arange(0,DeltaT.shape[0])*L/lam/np.shape(DeltaT)[0]
y = np.arange(0,DeltaT.shape[1])*L/lam/np.shape(DeltaT)[0]

# Putting origin in the center
#x += -np.max(x)/2
y# += -np.max(y)/2


# Plot model****************************************************
fig = plt.figure(figsize=(10, 6))
gs = gridspec.GridSpec(2, 2)

gs.update(wspace = 0.3,
         hspace = 0.25)

ax1 = plt.subplot(gs[:, 0])  # first column
ax2 = plt.subplot(gs[0, 1])  # second column, first row
ax3 = plt.subplot(gs[1, 1])  # second column, second row

## Vectorfield in Front of Colormap
im1 = ax1.pcolormesh(x,y,vp.T, cmap="plasma")

# Select every ... grid point for v representation
v_inc = 20 # increment: every incr-th v-value

# Create grid for plotting
x_every = x[::v_inc]
y_every = y[::v_inc]
# find fitting v-values
vx_every = vx[::v_inc, ::v_inc]
vy_every = vy[::v_inc, ::v_inc]



ax1.quiver(x_every,y_every,vx_every.T, vy_every.T, pivot = 'middle', color = 'lightsteelblue')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("P-Wave Velocity and flow Model")
cbar1 = fig.colorbar(im1, ax = ax1, fraction=0.046*(10/16), pad=0.04, label = '$c$')
ax1.set_aspect('equal')

## vx-Colormap
im2 = ax2.pcolormesh(x,y,vx.T, cmap="viridis")
cbar2 = fig.colorbar(im2, ax = ax2, fraction=0.046*(10/16), pad=0.04, label = '$vx$')
ax2.set_xlabel("x")
ax2.set_ylabel("y")

## vy-Colormap
im3 = ax3.pcolormesh(x,y,vy.T, cmap="viridis")
cbar3 = fig.colorbar(im3, ax = ax3, fraction=0.046*(10/16), pad=0.04, label = '$vy$')
ax3.set_xlabel("x")
ax3.set_ylabel("y")
#***************************************************************

plt.show()
```

Notice that the model is currently fairly coarse (ie. there are relatively few samples in the x- and y-directions).  This discretisation is currently too coarse for the desired source frequency we want to use for this example, so we need to interpolate this coarser model onto a finer grid.

Since the wavelength is fixed to $\lambda = 1$, we need the grid spacing `dh = L/lam/nx` to be desirably smaller than 0.1 - or even smaller since there is higher frequency content in each STF.

A fairly simple way of accomplishing this is to use the `RectBivariateSpline` tool from the `scipy` package.
We can set a desired scaling factor (see `scale`) which will scale the number of samples in each of the x- and y-directions.

```python
scale = 5
# routine to overwrite fields by more high resolved ones

# create spline object - for DeltaT
interp_DT = RectBivariateSpline(np.linspace(0, DeltaT.shape[0], DeltaT.shape[0]), np.linspace(0, DeltaT.shape[1], DeltaT.shape[1]), DeltaT)
# call the spline object on given gridpoints [in pt.-Number] - the larger 'scale', the finer the grid
DeltaT = interp_DT(np.linspace(0, DeltaT.shape[0], DeltaT.shape[0]*scale), np.linspace(0, DeltaT.shape[1], DeltaT.shape[1]*scale))

# repeat for vx
interp_vx = RectBivariateSpline(np.linspace(0, vx.shape[0], vx.shape[0]), np.linspace(0, vx.shape[1], vx.shape[1]), vx)
vx = interp_vx(np.linspace(0, vx.shape[0], vx.shape[0]*scale), np.linspace(0, vx.shape[1], vx.shape[1]*scale))

# repeat for vy
interp_vy = RectBivariateSpline(np.linspace(0, vy.shape[0], vy.shape[0]), np.linspace(0, vy.shape[1], vy.shape[1]), vy)
vy = interp_vy(np.linspace(0, vy.shape[0], vy.shape[0]*scale), np.linspace(0, vy.shape[1], vy.shape[1]*scale))

# rescale plotting grid
x = np.arange(0,DeltaT.shape[0])*L/lam/np.shape(DeltaT)[0]
y = np.arange(0,DeltaT.shape[1])*L/lam/np.shape(DeltaT)[0]
# update increment
v_inc *= scale

# CANCELLING THE DT FIELD
#******************************
#DeltaT *= 0
#******************************

# CANCELLING THE VELOCITY FIELD
#******************************
#vx *= 0
#vy *= 0
#vy_every *= 0
#vx_every *= 0
#******************************


# Creating the model array
m = np.zeros((3,np.shape(DeltaT)[0],np.shape(DeltaT)[1]))
m[0,:,:] = DeltaT
m[1,:,:] = vx
m[2,:,:] = vy
```

## Simulation Setup

Next, we can set up the simulation itself.  For this example, we will be explicitly defining the following:
- Time step
- Number of time steps
- Spatial discretisation
- `ijsrcs`: Indices of the source positions
- Source time function characteristics
- `ijrecs`: Indices of the receiver positions
- Free surface boundary conditions at the surface of the domain
- PMLs on all other sides

```python
# Temporal discretization
nt = 13000
dt = 0.005
t = np.arange(0.0, nt*dt, dt)

# Spatial discretization
nx = DeltaT.shape[0]
ny = DeltaT.shape[1]
dh = L/lam/nx

# Check grid dispersion and stability 
print('grid step: %.4f' %dh)
print('Number of Gridpoints %i x %i' %(nx,ny))
print("Stability criterion, CFL number:",(vp_max*dt/dh))

# Sources

# Selection of source positions
#ijsrc = np.array([[nx//2, 1]]).T  # -- one after another.
#ijsrc = np.array([[nx//2, ny-2]]).T

ijsrc_pool = np.array([[nx//5, 1],   # lower boundary
                  [2*nx//5, 1],
                  [3*nx//5, 1],
                  [4*nx//5, 1],
                  [nx//5, ny-2],  # upper boundary
                  [2*nx//5, ny-2],
                  [3*nx//5, ny-2],
                  [4*nx//5, ny-2],
                  [1, ny//5],   # left boundary
                  [1, 2*ny//5],
                  [1, 3*ny//5],
                  [1, 4*ny//5],
                  [nx-2, ny//5],  # right boundary
                  [nx-2, 2*ny//5],
                  [nx-2, 3*ny//5],
                  [nx-2, 4*ny//5]]).T

# Extracting the source positions
srcpos = ijsrc_pool.T * dh

print(("Source positions:\n{}".format(srcpos)))

# Source time (center)
t0 = 1
# Source center frequency
f0 = 1

# Source time function - as an array
sourcetf_pool = np.zeros((ijsrc_pool.shape[1],nt))
for s in range(ijsrc_pool.shape[1]):
    # delay may be added to each sorce time
    sourcetf_pool[s,:] = sourcetimefuncs.rickersource(t, t0, f0)

# Show STF:
plt.figure(figsize = [5,4])
plt.title('Source Time Function')
plt.plot(t,sourcetf_pool[0,:])
plt.xlim(0,2*t0)
plt.xlabel('t')
plt.show()

# Receivers
ijrec = ijsrc_pool # receivers at same position at sources

#ijrec = np.array([[nx//2, 1],
#                  [nx//2, 181],
#                  [nx//2, 361],
#                  [nx//2, ny-362],
#                  [nx//2, ny-182],
#                  [nx//2, ny-2]]).T # seperate receivers

# Extracting receiver coordinates
nrec = ijrec.shape[1]
recpos = ijrec.T * dh

print(("Receiver positions:\n{}".format(recpos)))


# Initiallize input parameter dictionary
inpar = {}
inpar["ntimesteps"] = nt
inpar["nx"] = nx
inpar["nz"] = ny
inpar["dt"] = dt
inpar["dh"] = dh
inpar["T0"] = T0            # dimensional for CR_Water
inpar["DeltaT0"] = DeltaT0  # dimensional for CR_Water
inpar["A"]  = alpha*DeltaT0
inpar["Ma"] = v0/c0
inpar["savesnapshot"] = False
inpar["snapevery"] = 500
inpar["savefinal"] = True
inpar["v_inc"] = v_inc
inpar["freesurface"] = False
inpar["boundcond"] = "ReflBou" #"GaussTap", "PML", "GaussTap","ReflBou" - only the latter for now
inpar["constrel"] = CR  #'IdealGas' - 'Water' still in progress

# Plot the Model
plt.figure(figsize=[8, 6])
im = plt.pcolormesh(x,y,DeltaT.T, cmap="plasma")
plt.plot(ijsrc_pool[0, :]*dh, ijsrc_pool[1, :]*dh, "*w", markersize = 15, label = 'Sources')
plt.plot(ijrec[0, :]*dh, ijrec[1, :]*dh, ".c", label = 'Receivers')
plt.quiver(x_every,y_every,vx_every.T, vy_every.T, pivot = 'middle', width = 0.0015, color = 'lightsteelblue')
plt.xlabel("x-Position")
plt.ylabel("y-Position")
plt.title("Flow and Temperature Model")
plt.colorbar(im, fraction=0.046*(10/16), pad=0.04, label = '$\Delta T$')
plt.gca().set_aspect('equal')
plt.legend(loc = 'center left')
plt.show()
```

## Running the Simulation - for several sources

Now that we have set up the forward simulation, we can run this using `solveacoustic2D` in `pestoseis.USwaveprop2D`. (modified version)

```python
folder = '../../Full-Waveform Data/Cyc_Test_Results/Phase II/Cyclone 13k_timesteps'
for s in [1,4]:
    #if s == 4:
    #    inpar["savesnapshot"] = True
    #else:
    #    inpar["savesnapshot"] = False
        
    ijsrc = ijsrc_pool[:,s].reshape(2,1)
    sourcetf = sourcetf_pool[s,:].reshape(1,nt)
    seism, psave = USwaveprop2D_on_cuda.solveacoustic2D(inpar, ijsrc, m, sourcetf, f0, ijrec,
                                                        outfileh5 = folder+'/Data_src-'+str(s)+'.h5')
```
