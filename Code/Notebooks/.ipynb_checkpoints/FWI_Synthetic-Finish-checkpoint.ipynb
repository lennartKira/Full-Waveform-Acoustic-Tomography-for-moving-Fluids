# Synthetic Test for Acoustic Velocimetry with Full-Waveform Inversion
A synthetic forward wavefield is simulated and recorded at certain receiver positions.\
Then a simulation is run using a prior model and the field is sampled at the same receiver positions. Additionally, the final conditions of the wavefield are stored.\
Finally, the adjoint simulation is simultaniously performed with a reverse simulation of the prior wavefield to create the gradient.

## Model Interpolation
Notice that the model is currently fairly coarse (ie. there are relatively few samples in the x- and y-directions).  This discretisation is currently too coarse for the desired source frequency we want to use for this example, so we need to interpolate this coarser model onto a finer grid.

Since the wavelength is fixed to $\lambda = 1$, we need the grid spacing `dh = L/lam/nx` to be desirably smaller than 0.1 - or even smaller since there is higher frequency content in each STF.

A fairly simple way of accomplishing this is to use the `RectBivariateSpline` tool from the `scipy` package.
We can set a desired scaling factor (see `scale`) which will scale the number of samples in each of the x- and y-directions.

```python
scale = 5
# routine to overwrite fields by more high resolved ones

# create spline object - for DeltaT
interp_DT = RectBivariateSpline(np.linspace(0, DeltaT.shape[0], DeltaT.shape[0]), np.linspace(0, DeltaT.shape[1], DeltaT.shape[1]), DeltaT)
# call the spline object on given gridpoints [in pt.-Number] - the larger 'scale', the finer the grid
DeltaT = interp_DT(np.linspace(0, DeltaT.shape[0], DeltaT.shape[0]*scale), np.linspace(0, DeltaT.shape[1], DeltaT.shape[1]*scale))

# repeat for vx
interp_vx = RectBivariateSpline(np.linspace(0, vx.shape[0], vx.shape[0]), np.linspace(0, vx.shape[1], vx.shape[1]), vx)
vx = interp_vx(np.linspace(0, vx.shape[0], vx.shape[0]*scale), np.linspace(0, vx.shape[1], vx.shape[1]*scale))

# repeat for vy
interp_vy = RectBivariateSpline(np.linspace(0, vy.shape[0], vy.shape[0]), np.linspace(0, vy.shape[1], vy.shape[1]), vy)
vy = interp_vy(np.linspace(0, vy.shape[0], vy.shape[0]*scale), np.linspace(0, vy.shape[1], vy.shape[1]*scale))

# rescale plotting grid
x = np.arange(0,DeltaT.shape[0])*L/lam/np.shape(DeltaT)[0]
y = np.arange(0,DeltaT.shape[1])*L/lam/np.shape(DeltaT)[0]
# update increment
v_inc *= scale

# CANCELLING THE DT FIELD
#******************************
#DeltaT *= 0
#******************************

# CANCELLING THE VELOCITY FIELD
#******************************
#vx *= 0
#vy *= 0
#vy_every *= 0
#vx_every *= 0
#******************************


# Creating the model array
m = np.zeros((3,np.shape(DeltaT)[0],np.shape(DeltaT)[1]))
m[0,:,:] = DeltaT
m[1,:,:] = vx
m[2,:,:] = vy
```

## Simulation Setup

Next, we can set up the simulation itself.  For this example, we will be explicitly defining the following:
- Time step
- Number of time steps
- Spatial discretisation
- `ijsrcs_pool`: Pool of indices of the source positions - from which each source is later chosen to emit
- Source time function characteristics
- `ijrecs`: Indices of the receiver positions
- Boundary Conditions and save options

```python
# Temporal discretization
dt = 0.005
nt = 14000
t = np.arange(0.0, nt*dt, dt)

# Spatial discretization
nx = DeltaT.shape[0]
ny = DeltaT.shape[1]
dh = L/lam/nx

# Check grid dispersion and stability 
print('grid step: %.4f' %dh)
print('Number of Gridpoints %i x %i' %(nx,ny))
print("Stability criterion, CFL number:",(vp_max*dt/dh))
###############################################

# Sources

# Selection of source positions
#ijsrc = np.array([[nx//2, 1]]).T  # -- one after another.
#ijsrc = np.array([[nx//2, ny-2]]).T

ijsrc_pool = np.array([[nx//5, 1],   # lower boundary
                  [2*nx//5, 1],
                  [3*nx//5, 1],
                  [4*nx//5, 1],
                  [nx//5, ny-2],  # upper boundary
                  [2*nx//5, ny-2],
                  [3*nx//5, ny-2],
                  [4*nx//5, ny-2],
                  [1, ny//5],   # left boundary
                  [1, 2*ny//5],
                  [1, 3*ny//5],
                  [1, 4*ny//5],
                  [nx-2, ny//5],  # right boundary
                  [nx-2, 2*ny//5],
                  [nx-2, 3*ny//5],
                  [nx-2, 4*ny//5]]).T

# Extracting the source positions
srcpos = ijsrc_pool.T * dh

print(("Source positions:\n{}".format(srcpos)))

###############################################
# Source time (center)
t0 = 1
# Source center frequency
f0 = 1
# Width of signal (cycles)
#wSTF = 4  # Required for sourcetimefuncs.cossource

# Source time function - as an array
sourcetf_pool = np.zeros((ijsrc_pool.shape[1],nt))

#STF_delays = np.random.rand(ijsrc_pool.shape[1])

for s in range(ijsrc_pool.shape[1]):
    # delay to t0 may be added to vary source times
    #************
    STF_delay = 0.2*s
    STF_scale = (1.5+0.1*(s-8))
    #************
    sourcetf_pool[s,:] = STF_scale*sourcetimefuncs.rickersource(t, t0+STF_delay, f0)

# Show STF:
plt.figure(figsize = [5,4])
plt.title('Source Time Function')
plt.plot(t,sourcetf_pool[2,:])
plt.xlim(0,5*t0)
plt.xlabel('t')
plt.show()
###############################################

# Receivers
ijrec = ijsrc_pool # receivers at same position at sources

#ijrec = np.array([[nx//2, 1],
#                  [nx//2, 181],
#                  [nx//2, 361],
#                  [nx//2, ny-362],
#                  [nx//2, ny-182],
#                  [nx//2, ny-2]]).T # seperate receivers

# Extracting receiver coordinates
nrec = ijrec.shape[1]
recpos = ijrec.T * dh

print(("Receiver positions:\n{}".format(recpos)))

###############################################
# Initiallize input parameter dictionary
inpar = {}
inpar["ntimesteps"] = nt
inpar["nx"] = nx
inpar["nz"] = ny
inpar["dt"] = dt
inpar["dh"] = dh
inpar["T0"] = T0            # dimensional for CR_Water
inpar["DeltaT0"] = DeltaT0  # dimensional for CR_Water
inpar["A"]  = alpha*DeltaT0
inpar["Ma"] = v0/c0
inpar["savesnapshot"] = False
inpar["snapevery"] = 1000
inpar["savefinal"] = False
inpar["v_inc"] = v_inc
inpar["freesurface"] = False
inpar["boundcond"] = "ReflBou" #"GaussTap", "PML", "GaussTap","ReflBou" - only the latter for now
inpar["constrel"] = CR  #'IdealGas' - 'Water' still in progress

###############################################################
# Finally, plot the Model including source-receiver-array
plt.figure(figsize=[8, 6])
im = plt.pcolormesh(x,y,DeltaT.T, cmap="plasma")
plt.plot(ijsrc_pool[0, :]*dh, ijsrc_pool[1, :]*dh, "*w", markersize = 15, label = 'Sources')
plt.plot(ijrec[0, :]*dh, ijrec[1, :]*dh, ".c", label = 'Receivers')
plt.quiver(x[::v_inc],y[::v_inc],vx[::v_inc, ::v_inc].T, vy[::v_inc, ::v_inc].T,
           pivot = 'middle', width = 0.0015, color = 'lightsteelblue')
plt.xlabel("x-Position")
plt.ylabel("y-Position")
plt.title("Flow and Temperature Model")
plt.colorbar(im, fraction=0.046*(10/16), pad=0.04, label = '$\Delta T$')
plt.gca().set_aspect('equal')
plt.legend(loc = 'center left')
plt.show()
```

```python
# free up memory
#grad = []

def L2_misfit(p, p0):
    f = 1/2*np.sum((p-p0)**2)*dt
    return f

# Initialize Misfit History
Chi_hist = np.array([])

# initialize prior model
m_prior = np.zeros((3,DeltaT.shape[0],DeltaT.shape[1]))

# load prior model and data
#h5_data = prior_folder+"/Data_src-all_dt0-02s_STFscaled.h5"
h5_data = folder+'/Data_src-all_m3_smooth.h5'
print('Prior Model Data:', h5_data)
h=h5py.File(h5_data,"r")

# import prior model
#----------------------------
m_prior[0,:,:] = h["DeltaT"][:,:].copy()
m_prior[1,:,:] = h["vx"][:,:]
m_prior[2,:,:] = h["vy"][:,:]

dt = h["dt"][()]
dh = h["dh"][()]
nx = m_prior[0,:,:].shape[0]
ny = m_prior[0,:,:].shape[1]

# import prior waveforms
#----------------------------
seism = h["seism"][:,:].copy()
h.close()

# load true data and compute misfit
h5_data = folder+"/Data_src-all_3K_dt0-02s_STFscaled.h5"
print('True Data:', h5_data)
h=h5py.File(h5_data,"r")

# import true waveforms
#----------------------
seism0 = h["seism"][:,:].copy()
h.close()

# Compute Misfit
Chi_0 = L2_misfit(seism, seism0)
Chi_hist = np.append(Chi_hist, Chi_0)
print('Initial Misfit:', Chi_0)

## load gradient
grad_old = np.zeros((3,nx,ny))
#h5_data = folder+"/Gradient_filtered_src-all_3K_dt0-02s_STFscaled.h5"
h5_data = folder+"/Gradient_filtered_m3_smooth.h5"
print('Gradient from:', h5_data)
h=h5py.File(h5_data,"r")

grad_old[0,:,:] = h["gradient_c"][:,:].copy()
grad_old[1,:,:] = h["gradient_vx"][:,:].copy()
grad_old[2,:,:] = h["gradient_vy"][:,:].copy()
h.close()

# Initialise arrays for storing curvature information:
#w_c = np.array([], dtype = float).reshape((nx,ny,0))
#s_c = np.array([], dtype = float).reshape((nx,ny,0))
#w_vx = np.array([], dtype = float).reshape((nx,ny,0))
#s_vx = np.array([], dtype = float).reshape((nx,ny,0))
#w_vy = np.array([], dtype = float).reshape((nx,ny,0))
#s_vy = np.array([], dtype = float).reshape((nx,ny,0))

# Load Prior Curvature Information
h5_data = folder + "/Model_iter-3_smooth.h5"
print('s&w from:', h5_data)
h=h5py.File(h5_data,"r")

w_c = h["w_c"][:,:].copy()
s_c = h["s_c"][:,:].copy()
w_vx = h["w_vx"][:,:].copy()
s_vx = h["s_vx"][:,:].copy()
w_vy = h["w_vy"][:,:].copy()
s_vy = h["s_vy"][:,:].copy()
Chi_hist = h["Misfit_history"][:].copy()
h.close()

c_1 = 1e-4 # factor for first Wolfe Condition
c_2 = 0.8  # factor for second Wolfe Condition
m_max = 3  # maximum number of gradients considerated for Hessian

sigma_cut = 2/dh # number of gridpts at which updates are fixed to zero

# Initialize first iteration
m_old = m_prior
Chi_old = Chi_0

#############################
# For what should be inverted:
for_Temperature = [True, True, True, True, True, True, True, False, False]
for_Velocity = [False, False, False, False, False, True, True, True, True]
#############################

# Starting the L-BFGS:
for k in [4,5,6,7,8]:
    save_appendix = str(k+1)+"_smooth.h5"
    print("***")
    print("Iteration %i" %(k+1))
    print("***")
    
    # initialize new model
    m_new = np.zeros((3,DeltaT.shape[0],DeltaT.shape[1]))
    
    m_new[0,:,:] = m_old[0,:,:]
    m_new[1,:,:] = m_old[1,:,:]
    m_new[2,:,:] = m_old[2,:,:]
    
    if for_Temperature[k]:
        # Update in c-Space:
        m_new[0,:,:] = np.sqrt(c2(m_new[0,:,:],T0,DeltaT0))
        #-----------------------------------
        
        if nT_iter == 0:
            Deltac_max = np.sqrt(c2(1, T0, DeltaT0))-1
            gamma_c = Deltac_max/np.max(np.abs(grad_old[0,:,:]))
        else:
            gamma_c = np.sum(s_c[:,:,-1]*w_c[:,:,-1])/np.sum(w_c[:,:,-1]*w_c[:,:,-1])
    
        p_ck = L_BFGS.compute_stepdir(grad_old[0,:,:], s_c, w_c, gamma_c)#*Inverse_utils.rec_cutmask(nx, ny, ijrec, sigma_cut)
        alpha_ck = 1
        # update model
        m_new[0,:,:] += alpha_ck*p_ck
        
        # Transform back to DT-Space 
        m_new[0,:,:] = c_inv(m_new[0,:,:], T0, DeltaT0)
        #------------------------------------
        nT_iter +=1
        
    if for_Velocity[k]:
        if nV_iter == 0:
            gamma_vx = 1/np.max(np.abs(grad_old[1,:,:]))
            gamma_vy = 1/np.max(np.abs(grad_old[2,:,:]))
        else:
            gamma_vx = np.sum(s_vx[:,:,-1]*w_vx[:,:,-1])/np.sum(w_vx[:,:,-1]*w_vx[:,:,-1])
            gamma_vy = np.sum(s_vy[:,:,-1]*w_vy[:,:,-1])/np.sum(w_vy[:,:,-1]*w_vy[:,:,-1])
    
        # step in vx-space
        p_vxk = L_BFGS.compute_stepdir(grad_old[1,:,:], s_vx, w_vx, gamma_vx)#*Inverse_utils.rec_cutmask(nx, ny, ijrec, sigma_cut)
        alpha_vxk = 1
        #step in vy-space
        p_vyk = L_BFGS.compute_stepdir(grad_old[2,:,:], s_vy, w_vy, gamma_vy)#*Inverse_utils.rec_cutmask(nx, ny, ijrec, sigma_cut)
        alpha_vyk = 1
        
        # update model
        m_new[1,:,:] += alpha_vxk*p_vxk
        m_new[2,:,:] += alpha_vyk*p_vyk
        nV_iter += 1
        
    
    
    # Compute new seismograms and misfit:
    print('Computing Prior data for model', k+1)
    inpar["savefinal"] = True
    # Run the simulation with all sources firing
    seism_new, psave_new = USwaveprop2D_on_cuda.solveacoustic2D(inpar, ijsrc_pool, m_new, sourcetf_pool, f0, ijrec,
                                                                outfileh5 = folder+'/Data_src-all_m'+save_appendix)
    
    Chi_new = L2_misfit(seism_new, seism0)
    
    print('Misfit after iteration'+str(k+1)+': ', Chi_new)
        
    ## computing the adjoint source
    adj_src = Inverse_utils.adjointsource_L2(seism0, seism_new)
    
    # get final conditions:
    h5_data = folder+'/Data_src-all_m'+save_appendix
    h=h5py.File(h5_data,"r")   
    p_final = h["p_final"][:,:,:].copy()
    h.close()
    
    print('Computing Gradient data for model', k+1)
    # compute gradient
    grad_new = Inverse_utils.computegrad2D(inpar, ijrec, m_new, adj_src,
                                       p_final, saveh5=True,
                                       outfileh5=folder+"/Gradient_src-all_m"+save_appendix)
    
    # Gradient Post-Processing 
    grad_new[0,:,:] = Inverse_utils.filter_grad(grad_new[0,:,:], dh, a_min = 175, f_max = 2)
    grad_new[1,:,:] = Inverse_utils.filter_grad(grad_new[1,:,:], dh, a_min = 175, f_max = 2)
    grad_new[2,:,:] = Inverse_utils.filter_grad(grad_new[2,:,:], dh, a_min = 175, f_max = 2)
    
    

    # WOLFE CONDITIONS:
    if for_Temperature[k]:
        W0_c = Chi_new < Chi_old + c_1*alpha_ck*np.sum(grad_old[0,:,:]*p_ck)    # first Wolfe condition for c-update
        W1_c = np.sum(grad_new[0,:,:]*p_ck) >= c_2*np.sum(grad_old[0,:,:]*p_ck) # second Wolfe condition for c-update
    if for_Velocity[k]:
        W0_vx = Chi_new < Chi_old + c_1*alpha_vxk*np.sum(grad_old[1,:,:]*p_vxk)    # first Wolfe condition for vx-update
        W1_vx = np.sum(grad_new[1,:,:]*p_vxk) >= c_2*np.sum(grad_old[1,:,:]*p_vxk) # second Wolfe condition for vx-update
        W0_vy = Chi_new < Chi_old + c_1*alpha_vyk*np.sum(grad_old[2,:,:]*p_vyk)    # first Wolfe condition for vy-update
        W1_vy = np.sum(grad_new[2,:,:]*p_vyk) >= c_2*np.sum(grad_old[2,:,:]*p_vyk) # second Wolfe condition for vy-update
    
    #####################################################################
    # Reiterate step length and recalculate gradients while not satisfied
    while (for_Temperature[k] and not (W0_c and W1_c)) or (for_Velocity[k] and not (W0_vx and W0_vy and W1_vx and W1_vy)):
        
        if for_Temperature[k]:
            # first Wolfe Condition:
            if not W0_c:
                print('First Wolfe Condition not satisfied for a_c', k)
                alpha_ck *= 0.4
        
            # Check second Wolfe Condition
            if not W1_c:
                print('Second Wolfe Condition not satisfied for a_c', k)
                alpha_ck *= 1.5
                
            # Update in c-Space:
            #-----------------------------------
            m_new[0,:,:] = np.sqrt(c2(m_old[0,:,:],T0,DeltaT0)) + alpha_ck*p_ck
            # Transform back to DeltaT-Space 
            m_new[0,:,:] = c_inv(m_new[0,:,:], T0, DeltaT0)
            #------------------------------------
                
        if for_Velocity[k]:
            # first Wolfe Condition:
            if not W0_vx:
                print('First Wolfe Condition not satisfied for a_vx', k)
                alpha_vxk *= 0.4
            if not W0_vy:
                print('First Wolfe Condition not satisfied for a_vy', k)
                alpha_vyk *= 0.4
        
            # Check second Wolfe Condition
            if not W1_vx:
                print('Second Wolfe Condition not satisfied for a_vx', k)
                alpha_vxk *= 1.5
            if not W1_vy:
                print('Second Wolfe Condition not satisfied for a_vy', k)
                alpha_vyk *= 1.5 

            # Update flow model
            m_new[1,:,:] = m_old[1,:,:] + alpha_vxk*p_vxk
            m_new[2,:,:] = m_old[2,:,:] + alpha_vyk*p_vyk
        

        # Compute new seismograms and misfit:
        print('Recomputing Prior data for model', k+1)
        inpar["savefinal"] = True
        # Run the simulation with all sources firing
        seism_new, psave_new = USwaveprop2D_on_cuda.solveacoustic2D(inpar, ijsrc_pool, m_new, sourcetf_pool, f0, ijrec,
                                                                    outfileh5 = folder+'/Data_src-all_m'+save_appendix)

        # get final conditions:
        h5_data = folder+'/Data_src-all_m'+str(k+1)+'_smooth.h5'
        h=h5py.File(h5_data,"r")   
        p_final = h["p_final"][:,:,:].copy()
        h.close()
        
        Chi_new = L2_misfit(seism_new, seism0)
        
        print('New misfit after iteration'+str(k+1)+': ', Chi_new)
            
        ## computing the adjoint source
        adj_src = Inverse_utils.adjointsource_L2(seism0, seism_new)
        
        print('Recomputing Gradient data for model', k+1)
        # compute gradient
        grad_new = Inverse_utils.computegrad2D(inpar, ijrec, m_new, adj_src,
                                           p_final, saveh5=True,
                                           outfileh5=folder+"/Gradient_src-all_m"+save_appendix)
        
        # Gradient Post-Processing 
        grad_new[0,:,:] = Inverse_utils.filter_grad(grad_new[0,:,:], dh, a_min = 175, f_max = 2)
        grad_new[1,:,:] = Inverse_utils.filter_grad(grad_new[1,:,:], dh, a_min = 175, f_max = 2)
        grad_new[2,:,:] = Inverse_utils.filter_grad(grad_new[2,:,:], dh, a_min = 175, f_max = 2)

        # Re-Evaluate Wolfe Conditions:
        if for_Temperature[k]:
            W0_c = Chi_new < Chi_old + c_1*alpha_ck*np.sum(grad_old[0,:,:]*p_ck)    # first Wolfe condition for c-update
            W1_c = np.sum(grad_new[0,:,:]*p_ck) >= c_2*np.sum(grad_old[0,:,:]*p_ck) # second Wolfe condition for c-update
        if for_Velocity[k]:
            W0_vx = Chi_new < Chi_old + c_1*alpha_vxk*np.sum(grad_old[1,:,:]*p_vxk)    # first Wolfe condition for vx-update
            W1_vx = np.sum(grad_new[1,:,:]*p_vxk) >= c_2*np.sum(grad_old[1,:,:]*p_vxk) # second Wolfe condition for vx-update
            W0_vy = Chi_new < Chi_old + c_1*alpha_vyk*np.sum(grad_old[2,:,:]*p_vyk)    # first Wolfe condition for vy-update
            W1_vy = np.sum(grad_new[2,:,:]*p_vyk) >= c_2*np.sum(grad_old[2,:,:]*p_vyk) # second Wolfe condition for vy-update
    #####################################################################

    # Save last Misfit
    Chi_hist = np.append(Chi_hist, Chi_new)
    
    # append w and s
    if for_Temperature[k]:
        w_c = np.dstack((w_c, grad_new[0,:,:] - grad_old[0,:,:]))
        s_c = np.dstack((s_c, np.sqrt(c2(m_new[0,:,:], T0, DeltaT0)) 
                       - np.sqrt(c2(m_old[0,:,:], T0, DeltaT0))))
        # delete too old information:
        if nT_iter >= m_max:
            w_c = w_c[:,:,1:]
            s_c = s_c[:,:,1:]
            
    if for_Velocity[k]:
        w_vx = np.dstack((w_vx, grad_new[1,:,:] - grad_old[1,:,:]))
        s_vx = np.dstack((s_vx, m_new[1,:,:] - m_old[1,:,:]))
        w_vy = np.dstack((w_vy, grad_new[2,:,:] - grad_old[2,:,:]))
        s_vy = np.dstack((s_vy, m_new[2,:,:] - m_old[2,:,:]))
        # delete too old information:
        if nV_iter >= m_max:
            w_vx = w_vx[:,:,1:]
            s_vx = s_vx[:,:,1:]
            w_vy = w_vy[:,:,1:]
            s_vy = s_vy[:,:,1:]
        

    # save model progress
    outfileh5 = folder + "/Model_iter-"+save_appendix
    hf = h5py.File(outfileh5,"w")
    hf["DeltaT"] = m_new[0,:,:]
    hf["vx"] = m_new[1,:,:]
    hf["vy"] = m_new[2,:,:]
    hf["w_c"] = w_c
    hf["s_c"] = s_c
    hf["w_vx"] = w_vx
    hf["s_vx"] = s_vx
    hf["w_vy"] = w_vy
    hf["s_vy"] = s_vy
    hf["Misfit_history"] = Chi_hist
    hf.close()
    print("Saved Model and Curvature Info to ",outfileh5)
        
    # update model
    m_old = m_new
    Chi_old = Chi_new
    grad_old = grad_new
```
