## Model Specifications

```python
# typical temperature anomaly
DeltaT0 = 10 # [K]
# Background Temperature
T0 = 300     # [K]
# Typical Flow Speed
v0 = 1e-1  # [m/s]
# typical lengthscale
L = 0.5      # [m]
# sound wavelength
lam = 1e-2

# Constituive Relationship
CR = 'IdealGas' #'IdealGas' - 'Water' still in progress

# Choose Density effects to be included
Density_effects = False # Set to True for density effects

# Import relevant CR:
sys.path.append(
    '../ConstituiveRelationships')

if CR == 'IdealGas':
    from CR_IdealGas import c2
    # expansion coeff
    alpha = 1/T0 # [1/K] might be non-const. for DeltaT0/T0->1
    # typical soundspeed
    c0 = 300     # [m/s]
        
elif CR == 'Water':
    from CR_Water import c2
    from CR_Water import c2_dim
    # expansion coefficient
    alpha = 2e-4 # [1/K] Putintsev 2005
    # typical soundspeed
    c0 = np.sqrt(c2_dim(DeltaT, T0, DeltaT0)) # [m/s]        
else:
    print('wrong constituive relationship')
    
print('Vol. Expansion Coeff. %.1e 1/K' %alpha)

if not Density_effects:
    alpha = 0
    
    
print('Background Sound Speed: %i m/s' %c0)   # [m/s]
print('Mach Number: %.1e' %(v0/c0))
print('************')
print('Order of Magnitude Phaseshift [°]:', 2*L*v0/(c0*lam)*360)
```

## Import Waveforms (and information on orgininal model)

```python
folder = '../../Full-Waveform Data/Wiens_Test_Results/Phase II/13k_timesteps'


#Select Data
#h5_data = folder+"/Data_src-all.h5"

# Select Data (via source number)
s = 0
h5_data = folder+"/Data_src-"+str(s)+".h5"


print('Filename:', h5_data)
h=h5py.File(h5_data,"r")
# import model information
#----------------------------
sourcetf = h["srctf"][:].copy()
dt = h["dt"][()]
dh = h["dh"][()]
nx = h["nx"][()]
ny = h["nz"][()]
DeltaT = h["DeltaT"][:,:].copy()
vx = h["vx"][:,:]
vy = h["vy"][:,:]
v_inc = h["v_inc"][()]
ijrec = h["recij"][:,:].copy()
ijsrc = h["srcij"][:,:].copy()

f0 = 1.

# import true waveforms
#----------------------
#p_final = h["p_final"][:,:,:].copy()
seism0 = h["seism"][:,:].copy()

# import snapshots
#-----------------
#press = h["press"][:,:,:].copy()
#snapevery = h["snapevery"][()]
h.close()
```

## Prior Model

```python
# constructing the prior model
m_prior = np.zeros((3,DeltaT.shape[0],DeltaT.shape[1]))
x = np.linspace(0,(nx-1)*dh, nx)
y = np.linspace(0,(ny-1)*dh, ny)

# Plot the prior Model
plt.figure(figsize=[8, 6])
im = plt.pcolormesh(x,y,m_prior[0,:,:].T, cmap="plasma")
plt.plot(ijsrc[0, :]*dh, ijsrc[1, :]*dh, "*w", markersize = 15, label = 'Sources')
plt.plot(ijrec[0, :]*dh, ijrec[1, :]*dh, ".c", label = 'Receivers')
#plt.quiver(x[::v_inc],y[::v_inc],m_prior[1,::v_inc,::v_inc].T, m_prior[2,::v_inc,::v_inc].T, pivot = 'middle', width = 0.0015, color = 'lightsteelblue')
plt.xlabel("x-Position")
plt.ylabel("y-Position")
plt.title("Flow and Temperature Model")
plt.colorbar(im, fraction=0.046*(10/16), pad=0.04, label = '$\Delta T$')
plt.gca().set_aspect('equal')
plt.legend(loc = 'center left')
plt.show()
```

```python
# Initiallize input parameter dictionary
nt = sourcetf.shape[1]
inpar = {}
inpar["ntimesteps"] = nt #len(sourcetf) # for older models
inpar["nx"] = nx
inpar["nz"] = ny
inpar["dt"] = dt
inpar["dh"] = dh
inpar["T0"] = T0            # dimensional for CR_Water
inpar["DeltaT0"] = DeltaT0  # dimensional for CR_Water
inpar["A"]  = alpha*DeltaT0
inpar["Ma"] = v0/c0
inpar["savesnapshot"] = False
inpar["snapevery"] = 1000
inpar["savefinal"] = True
inpar["v_inc"] = v_inc
inpar["freesurface"] = False
inpar["boundcond"] = "ReflBou" #"GaussTap", "PML", "GaussTap","ReflBou" - only the latter for now
inpar["constrel"] = CR  #'IdealGas' - 'Water' still in progress
```
