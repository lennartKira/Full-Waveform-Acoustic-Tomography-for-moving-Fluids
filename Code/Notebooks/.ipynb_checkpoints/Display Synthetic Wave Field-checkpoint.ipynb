# Display Wave Field infront of the Flow Model
This Notebook loads snapshots of the synthetic wavefields and displays them as a video.\
Additionally, a snapshot at a selected timestamp can be displayed.

```python
#import standard packages
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.size"] = 14
```

```python
# Import specific packages
import h5py
from scipy.interpolate import RectBivariateSpline
from scipy.signal import correlate
import scipy.special
import sys
from matplotlib import gridspec 
import os # for file administration

# Add the folder containing .py files to the Python path
sys.path.append(
    '../Scripts')

#import Custom Packages
import AnimateWaves
import Plotting
```

```python
# Define the folder where to get the wave field from
folder = '../../Data/Full-Waveform Data and Reconstructions/Precession 4x4'

# load file
h5_data = folder+"/Data_src-[0]_rec-[7]_mediumSTF.h5"

h=h5py.File(h5_data,"r")
press = h["press"][:,:,:].copy()
srctf = h["sourcetf"][:].copy()
dt = h["dt"][()]
dh = h["dh"][()]
nx = h["nx"][()]
ny = h["ny"][()]
DeltaT = h["DeltaT"][:,:]
vx = h["vx"][:,:]
vy = h["vy"][:,:]
seism0 = h["seism"][:,:].copy()
snapevery = h["snapevery"][()]
comp_fact = h["comp_fact"][()]
ijrec = h["recij"][:,:].copy()
ijsrc = h["srcij"][:,:].copy()
h.close()
```

## Animate Forward Run

Start the animation:

```python
anim = AnimateWaves.animateacousticwaves(h5_data, fps = 20, clipamplitude=0.2, outfile = folder+"/animation.mp4")
plt.close(anim._fig)
```

Show the Animation in this Notebook

```python
from IPython.display import HTML
HTML(anim.to_html5_video())
```

## Visualise a single Snapshot

```python
x = np.arange(0,nx)*dh
y = np.arange(0,ny)*dh

# snapshot index
n_snapshot = 200 
clipamplitude=0.2

Plotting.PlotSnapshot_WaveField(x,y,DeltaT,vx,vy,press,comp_fact,ijrec,ijsrc,n_snapshot,clipamplitude, v_inc = 10)
```

## Visualise a single Waveform

```python
t = np.arange(srctf.shape[1])*dt

plt.figure(figsize = [6,2.5])

FA_slice = np.where(t<70)[0]  # slice of indeces for first arrival
SA_slice = np.where(t>=70)[0] # slice of indeces for secondary arrivals
plt.plot(t[FA_slice], seism0[0,FA_slice], linewidth = 1, color = 'darkblue')
plt.plot(t[SA_slice], seism0[0,SA_slice], linewidth = 1, color = 'lightgreen')
plt.vlines([70], ymin = -0.2, ymax = 0.2, color = "lightgrey", linestyle = "--")
plt.xlabel("$t/ \\tau_c$")
plt.ylabel("$p^\\text{obs}$ [Pa]")
plt.xticks(list(plt.xticks()[0]) + [70, t[-1]], labels = list(plt.xticks()[0].astype(str)) + ["$\\mathcal{T}_f$", "$\\mathcal{T}_r$"])
plt.xlim(0, t[-1])
plt.ylim((-0.11,0.11))
plt.savefig("../../Figures/WaveformExample.jpg", dpi = 300, bbox_inches = "tight")
```
