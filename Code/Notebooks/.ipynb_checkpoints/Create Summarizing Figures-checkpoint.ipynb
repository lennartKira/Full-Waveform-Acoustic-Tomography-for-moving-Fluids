# Generate the Summarizing Figures
This routine generates the central figures which are shown in the manuscript.\
The data is read from the `Results` folder where it has been stored using the `Evaluation` Manuscript.

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import sys
from matplotlib import gridspec

# Add the folder containing .py files to the Python path
sys.path.append(
    '../Scripts')

import SensorArrays
```

## Plot the Reconstruction Errors for all considered inversions:

```python
# Read all Errors from Results File:
Arrays = ["4x4", "4x2", "2x2"]

# velocity:
Models = ["Precession", "BuoyancyFree"]
RRSE_v_FWI = []
RRSE_v_SRI = []
for Model in Models:
    fname = f"../../Data/Results/{Model}_Wrapped.h5"
    with h5py.File(fname, "r") as hf:
        for Array in Arrays:
            RRSE_v_FWI.append(hf[f"v_Error_{Array}_FWI"][()])
            RRSE_v_SRI.append(hf[f"v_Error_{Array}_SRI"][()])

# Temperature:
Models = ["Wiens", "BuoyancyFree"]
RRSE_DT_FWI = []
RRSE_DT_SRI = []
for Model in Models:
    fname = f"../../Data/Results/{Model}_Wrapped.h5"
    with h5py.File(fname, "r") as hf:
        for Array in Arrays:
            RRSE_DT_FWI.append(hf[f"DeltaT_Error_{Array}_FWI"][()])
            RRSE_DT_SRI.append(hf[f"DeltaT_Error_{Array}_SRI"][()])

RRSE_v_FWI = np.array(RRSE_v_FWI)
RRSE_v_SRI = np.array(RRSE_v_SRI)
RRSE_DT_FWI = np.array(RRSE_DT_FWI)
RRSE_DT_SRI = np.array(RRSE_DT_SRI)
```

```python
# Set the global font size
plt.rcParams.update({'font.size': 14})
```

```python
plt.figure(figsize = (7,4))

plt.plot([3,4,5,6,7,8], RRSE_v_FWI, marker = "o", color = "lightgreen", linestyle = "", markersize = 10, label = "$v$ (FWI)")
plt.plot([0,1,2,6,7,8], RRSE_DT_FWI, marker = "o", color = "darkblue", linestyle = "", markersize = 9, label = "$\\Delta T$ (FWI)")
plt.plot([3,4,5,6,7,8], RRSE_v_SRI, marker = "d", color = "lightgreen", linestyle = "", markersize = 8, label = "$v$ (SRI)")
plt.plot([0,1,2,6,7,8], RRSE_DT_SRI, marker = "d", color = "darkblue", linestyle = "", markersize = 7, label = "$\\Delta T$ (SRI)")

#plt.xlim(-0.2,len(E_FWI_40x40)-0.8)
plt.ylim(0,)
plt.xticks([0,1,2,3,4,5,6,7,8], ["4x4","4x2","2x2","4x4","4x2","2x2","4x4","4x2","2x2"])

plt.vlines([2.5, 5.5], ymin = -0.05, ymax = 2.5, color = "grey", linestyle = "--")
plt.xlim(-0.5, 8.5)
plt.ylim(-0.05,2.3)
plt.text(-0.4,1.3, "Static Fluid", fontsize = 12)
plt.text(2.6,2.15, "Isothermal Flow", fontsize = 12)
plt.text(5.6,2.15, "Convective Flow", fontsize = 12)
plt.ylabel("$E$")
plt.legend(fontsize = 10, loc = "upper left", facecolor = "white")

#plt.savefig("../../Figures/Error_Summary.jpg", dpi = 300, bbox_inches = "tight")
plt.show()
```

## Plot all Reconstructions for a Static Temperature Anomaly

```python
# Adjust Font Size 
plt.rcParams.update({'font.size': 10.5})
```

```python
# Read File
fname = "../../Data/Results/Wiens_Wrapped.h5"

Arrays = ["4x4", "4x2", "2x2"]
Methods = ["FWI", "SRI"]
Fields = ["DeltaT"]

Reconstructions = []

with h5py.File(fname, "r") as hf:
    x = hf["x"][:]
    y = hf["y"][:]
    DeltaT = hf["DeltaT_true"][:,:]
    # read reconstructions
    
    for Array in Arrays:
        for Method in Methods:
            for Field in Fields:
                Reconstructions.append(hf[f"{Field}_{Array}_{Method}"][:,:])

SensorPos = []
ijsrc, ijrec = SensorArrays.Transducers4x4(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
ijsrc, ijrec = SensorArrays.Transducers4x2(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
ijsrc, ijrec = SensorArrays.Transducers2x2(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
```

```python
fig = plt.figure(figsize = (8,9))

cmap = "inferno"
gs = gridspec.GridSpec(3,3)

ax1 = plt.subplot(gs[0,0])

clim = [np.min(DeltaT), np.max(DeltaT)]

im1 = ax1.pcolormesh(x,y,DeltaT.T, cmap = cmap, clim = clim)
ax1.set_aspect("equal")
ax1.set_xticks([])
ax1.set_ylabel("$y$ [m]")
gs.update(wspace = 0.1, hspace = -0.1)

cax = fig.add_axes([ax1.get_position().x1 + 0.02,  # x position (right of ax_main)
                    ax1.get_position().y0,         # y position (same as ax_main)
                    0.02,  # width of colorbar
                    ax1.get_position().height])  # height of colorbar
fig.colorbar(im1, cax=cax, label = "$\\Delta T$ [K]")

data_idx = 0
for j in [0,1,2]:
    for i in [1,2]:
        data = Reconstructions[data_idx]
        ax = plt.subplot(gs[i,j])
        ax.pcolormesh(x,y,data.T, cmap = cmap, clim = clim)
        ax.scatter(SensorPos[j][0,:], SensorPos[j][1,:], marker = "X", c = "white", s = 35)
        data_idx += 1
        
        ax.set_xticks([]) if i != 2 else ax.set_xlabel("$x$ [m]")   
        ax.set_yticks([]) if j != 0 else ax.set_ylabel("$y$ [m]")
        
        ax.set_aspect("equal")

#plt.savefig("../../Figures/Wiens_Reconstructions.jpg", dpi = 500, bbox_inches = "tight")
plt.show()
```

## Plot all Reconstructions of the Isothermal Flow

```python
# Read File
fname = "../../Data/Results/Precession_Wrapped.h5"

Methods = ["FWI", "SRI"]
Arrays = ["4x4", "4x2", "2x2"]
Fields = ["vx", "vy"]

Reconstructions = []

with h5py.File(fname, "r") as hf:
    x = hf["x"][:]
    y = hf["y"][:]
    vx = hf["vx_true"][:,:]
    vy = hf["vy_true"][:,:]
    # read reconstructions
    for Method in Methods:
        for Array in Arrays:
            for Field in Fields:
                Reconstructions.append(hf[f"{Field}_{Array}_{Method}"][:,:])

SensorPos = []
ijsrc, ijrec = SensorArrays.Transducers4x4(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
ijsrc, ijrec = SensorArrays.Transducers4x2(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
ijsrc, ijrec = SensorArrays.Transducers2x2(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
```

```python
fig = plt.figure(figsize = (8,9))

cmap = "viridis"
v_inc = 8
gs = gridspec.GridSpec(3,3)

quiver_width = 0.005
quiver_scale = 0.5

gs.update(wspace = 0.1, hspace = -0.1)
ax1 = plt.subplot(gs[0,0])

clim = [np.min(np.sqrt(vx**2 + vy**2)), np.max(np.sqrt(vx**2 + vy**2))]

im1 = ax1.pcolormesh(x,y,np.sqrt(vx**2 + vy**2).T, cmap = cmap, clim = clim)
ax1.quiver(x[::v_inc],y[::v_inc],vx[::v_inc, ::v_inc].T, vy[::v_inc, ::v_inc].T,
           pivot = 'middle', color = 'snow', width = quiver_width, scale = quiver_scale)
ax1.set_aspect("equal")
ax1.set_xticks([])
ax1.set_ylabel("$y$ [m]")

cax = fig.add_axes([ax1.get_position().x1 + 0.02, ax1.get_position().y0, 0.015, ax1.get_position().height])
fig.colorbar(im1, cax=cax, label = "$u$ [m/s]")#, ticks = [0.01,0.04])

data_idx = 0
for i in [1,2]:
    for j in [0,1,2]:
        data_x = Reconstructions[data_idx]
        data_y = Reconstructions[data_idx+1]
        ax = plt.subplot(gs[i,j])
        ax.pcolormesh(x,y,np.sqrt(data_x**2 + data_y**2).T, cmap = cmap, clim = clim)
        ax.quiver(x[::v_inc],y[::v_inc],data_x[::v_inc, ::v_inc].T, data_y[::v_inc, ::v_inc].T,
                  pivot = 'middle', color = 'snow', width = quiver_width, scale = quiver_scale)
        ax.scatter(SensorPos[j][0,:], SensorPos[j][1,:], marker = "X", c = "white", s = 35)
        
        ax.set_xticks([]) if i != 2 else ax.set_xlabel("$x$ [m]")
        ax.set_yticks([]) if j != 0 else ax.set_ylabel("$y$ [m]")
            
        ax.set_aspect("equal")
        
        data_idx += 2

plt.savefig("../../Figures/Precession_Reconstructions.jpg", dpi = 500, bbox_inches = "tight")
plt.show()
```

## Plot the joint Reconstructions of Temperature and Flow

```python
# Read File
fname = "../../Data/Results/BuoyancyFree_Wrapped.h5"

Methods = ["FWI", "SRI"]
Arrays = ["4x4", "4x2", "2x2"]
Fields = ["DeltaT", "vx", "vy"]

Reconstructions = []

with h5py.File(fname, "r") as hf:
    x = hf["x"][:]
    y = hf["y"][:]
    DeltaT = hf["DeltaT_true"][:,:]
    vx = hf["vx_true"][:,:]
    vy = hf["vy_true"][:,:]
    # read reconstructions
    
    for Method in Methods:
        for Array in Arrays:
            for Field in Fields:
                Reconstructions.append(hf[f"{Field}_{Array}_{Method}"][:,:])

SensorPos = []
ijsrc, ijrec = SensorArrays.Transducers4x4(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
ijsrc, ijrec = SensorArrays.Transducers4x2(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
ijsrc, ijrec = SensorArrays.Transducers2x2(len(x),len(y))
SensorPos.append(ijsrc*(x[1]-x[0]))
```

```python
fig = plt.figure(figsize = (8,9))

cmap = "inferno"
v_inc = 8
quiver_width = 0.006
quiver_scale = 0.1
gs = gridspec.GridSpec(3,3)

ax1 = plt.subplot(gs[0,0])

clim = [np.min(DeltaT), np.max(DeltaT)]

im1 = ax1.pcolormesh(x,y,DeltaT.T, cmap = cmap, clim = clim)
im1_quiver = ax1.quiver(x[::v_inc],y[::v_inc],vx[::v_inc, ::v_inc].T, vy[::v_inc, ::v_inc].T,
                        pivot = 'middle', color = 'snow', width = quiver_width, scale = quiver_scale)
ax1.set_aspect("equal")
ax1.set_xticks([])
ax1.set_ylabel("$y$ [m]")
gs.update(wspace = 0.1, hspace = -0.1)

cax = fig.add_axes([ax1.get_position().x1 + 0.02, ax1.get_position().y0, 0.02, ax1.get_position().height])

fig.colorbar(im1, cax=cax, label = "$\\Delta T$ [K]")

#draw box to add the key on top
qk = plt.quiverkey(im1_quiver, 1.4, 0.02, U = 0.01, label = "1 cm/s", color = "black")

data_idx = 0
for i in [1,2]:
    for j in [0,1,2]:
        data = Reconstructions[data_idx]
        data_x = Reconstructions[data_idx+1]
        data_y = Reconstructions[data_idx+2]
        ax = plt.subplot(gs[i,j])
        ax.pcolormesh(x,y,data.T, cmap = cmap, clim = clim)
        ax.quiver(x[::v_inc],y[::v_inc],data_x[::v_inc, ::v_inc].T, data_y[::v_inc, ::v_inc].T,
                  pivot = 'middle', color = 'snow', width = quiver_width, scale = quiver_scale)
        ax.scatter(SensorPos[j][0,:], SensorPos[j][1,:], marker = "X", c = "white", s = 35)
        data_idx += 3
        
        ax.set_xticks([]) if i != 2 else ax.set_xlabel("$x$ [m]")   
        ax.set_yticks([]) if j != 0 else ax.set_ylabel("$y$ [m]")
        
        ax.set_aspect("equal")

#plt.savefig("../../Figures/BuyancyFree_Reconstructions.jpg", dpi = 500, bbox_inches = "tight")
plt.show()
```
