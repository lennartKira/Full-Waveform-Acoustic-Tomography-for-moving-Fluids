```python
def Test(x,y):
    
    f = np.zeros((len(x), len(y)))
    x_m = x[int(len(x)/2)]
    for i in range(len(x)):
        for j in range(len(y)):
            f[i,j] = 1 - ((x_m-x[i])/x_m)**2.
    
    return f
```

## Uniform Temperature distribution in a Rectangular Domain

```python
# grid point numbers
nx = 100
ny = 200

x = np.linspace(0,nx,nx)
y = np.linspace(0,ny,ny)

# temperature distribution
DeltaT = Test(x,y)/np.max(Test(x,y))
```

```python
plt.pcolormesh(x,y,DeltaT.T)
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label = '$\Delta$T')
```

### Saving the model as .h5-file

```python
# Create an HDF5 file in write mode ('w')
with h5py.File('DT_model_Test.h5', 'w') as f:
    f.create_dataset('DeltaT', data = DeltaT)
    f.create_dataset('vx', data = v[0,:,:])
    f.create_dataset('vy', data = v[1,:,:])
    f.create_dataset('ijsrcs', data = ijSources)
    f.create_dataset('ijrecs', data = ijReceivers)
```
