```python
import numpy as np
import matplotlib.pyplot as plt
import sys

# package for efficient array storage
import h5py
```

```python
def Cycl(x,y):
    
    v = np.zeros((2, len(x), len(y))) #2D Vector field (flow velocity)
    for i in range(len(x)):
        for j in range(len(y)):
            r = np.sqrt(x[i]**2 + y[j]**2.)
            v[0,i,j] = -y[j]
            v[1,i,j] = x[i]
    
    return v

def Gaussian(x,y,sigma):
    
    f = np.zeros((len(x), len(y)))
    for i in range(len(x)):
        for j in range(len(y)):
            f[i,j] = np.exp(-(((x[i]-np.mean(x))**2. + (y[j]-np.mean(y))**2.)/sigma**2.))/np.sqrt(2*np.pi*sigma**2.)
    
    return f
```

## Cyclonic Windfield: $v = r\,e_\varphi$

```python
# grid point numbers
nx = 512
ny = 512

# coordinates
x = np.arange(0,nx)
y = np.arange(0,ny)

# centered coordinates
xc = x - nx/2
yc = y - ny/2

# temperature distribution
v = Cycl(xc,yc)/np.max(Cycl(xc,yc))
```

```python
# Select every ... grid point for v representation
incr = 5 # increment: every ... v-value

vx_every = v[0, ::incr, ::incr]
vy_every = v[1, ::incr, ::incr]

# Create grid for plotting
x_every = x[::incr]
y_every = y[::incr]
```

## Gaussian Temperature Field

```python
# width of anomaly
sigma = nx/2

# temperature distribution
DeltaT = Gaussian(xc,yc, sigma)/np.max(Gaussian(xc,yc, sigma))
DeltaT = (DeltaT-0.5)*2
```

## Display Model

```python
plt.figure()
plt.pcolormesh(x,y,DeltaT)
plt.quiver(x_every,y_every,vx_every.T, vy_every.T, color = 'white', pivot = 'middle')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Saving the model as .h5-file

```python
# Create an HDF5 file in write mode ('w')
with h5py.File('cyclone_model.h5', 'w') as f:
    f.create_dataset('DeltaT', data = DeltaT)
    f.create_dataset('vx', data = v[0,:,:])
    f.create_dataset('vy', data = v[1,:,:])
```

## Setting Sources and Receiver Positions

```python
ijsrc = np.array([[nx//5, 1],   # lower boundary
                  [2*nx//5, 1],
                  [3*nx//5, 1],
                  [4*nx//5, 1],
                  [nx//5, ny-2],  # upper boundary
                  [2*nx//5, ny-2],
                  [3*nx//5, ny-2],
                  [4*nx//5, ny-2],
                  [1, ny//5],   # right boundary
                  [1, 2*ny//5],
                  [1, 3*ny//5],
                  [1, 4*ny//5],
                  [nx-2, ny//5],  # left boundary
                  [nx-2, 2*ny//5],
                  [nx-2, 3*ny//5],
                  [nx-2, 4*ny//5]]).T # source postions

ijrec = ijsrc # sources and receivers coincide
```

## Display Source and receiver positions

```python
plt.figure(figsize = [7,7])
plt.plot(ijsrc[0, :], ijsrc[1, :], "*r", markersize = 10, label = 'Sources')
plt.plot(ijrec[0, :], ijrec[1, :], ".c", label = 'Receivers')
plt.pcolormesh(x,y,DeltaT)
plt.quiver(x_every,y_every,vx_every.T, vy_every.T, color = 'white', pivot = 'middle')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(bbox_to_anchor=(1, 0.5))
plt.gca().set_aspect('equal')
plt.show()
```
