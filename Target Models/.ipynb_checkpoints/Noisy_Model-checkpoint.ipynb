```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
import scipy.io
import h5py
```

```python
loc = '../../wiens_model/'
```

```python
data = scipy.io.loadmat(loc+'wiens_model_512x512_Temp.mat')
```

## Load flow field of Wiens

```python
# reading grid locations
x = data['x_plot2'].flatten()
y = data['y_plot2'].flatten()

# reading velocity components
v_x = data['u_rbf2'].reshape((len(x),len(y)))
v_y = data['v_rbf2'].reshape((len(x),len(y)))

# plot absolute flow speed
v = np.sqrt(v_x**2 + v_y**2)

print('Maximum Flow Speed: ', np.max(v))

# Normalization:
v_x = v_x/np.max(v)
v_y = v_y/np.max(v)
```

```python
plt.figure()
plt.pcolormesh(x, y, v.T)

# plot quiver
step = 16

plt.quiver(x[::step], y[::step], v_x.T[::step,::step], v_y.T[::step,::step], color = 'white')
```

## Random Temperature Anomaly

```python
# Parameters
domain_size = 50  # in cm
grid_size = len(x)
anomaly_size = 0.5*grid_size/domain_size 

# Generate random field with values between 0 and 1
random_field = np.random.rand(grid_size, grid_size)

# Scale and shift to get values between -1 and 1
random_field = 2 * random_field - 1

# Apply Gaussian filter to smooth the field
sigma = anomaly_size  # standard deviation for Gaussian kernel
smooth_field = gaussian_filter(random_field, sigma=sigma)

# Normalize the field to the range [0, 1] for visualization
smooth_field = (smooth_field ) / np.max(np.abs(smooth_field))

# Plot the field
plt.imshow(smooth_field, cmap='viridis', extent=(0, domain_size, 0, domain_size))
plt.colorbar(label='Scalar Field Value')
plt.title('Gaussian Random Field with Irregular Anomalies')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()

DeltaT = smooth_field
```

## Save Model as h5-file

```python
# Create an HDF5 file in write mode ('w')
with h5py.File('Noisy_model.h5', 'w') as f:
    f.create_dataset('DeltaT', data = DeltaT)
    f.create_dataset('vx', data = v_x[:,:])
    f.create_dataset('vy', data = v_y[:,:])
```
