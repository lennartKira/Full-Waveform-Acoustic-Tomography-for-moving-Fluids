```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
import scipy.io
import h5py
```

## Define Base Flow of Libration

```python
nx = 100
ny = 100

L = 0.25 # [m] half-hight of cuboid

x = np.linspace(-L,L,nx)
y = np.linspace(-L,L,ny)

# Defining Dimensions
x_mesh, y_mesh = np.meshgrid(x, y) #[m]



# Libration Characteristics
eps = 0.3 # libration angle
Omega0 = 4 # [rad/s] - Angular Velocity

# Defining the Streamfunction
# Infinite Series in Solution:
xL = x_mesh/L
yL = y_mesh/L

def d(n):
    f = (2*n+1)*np.pi/2
    return f

inf_sr = 0
for n in range(100):
    inf_sr += (-1)**n*(np.cos(d(n)*xL)*np.cosh(d(n)*yL) + np.cos(d(n)*yL)*np.cosh(d(n)*xL))/(d(n)**3*np.cosh(d(n)))


Psi = eps*Omega0*(1 - 0.5*(xL**2 + yL**2) - 2*inf_sr)

# Obtaining the velocity field:
vx = -np.gradient(Psi, axis = 1)
vy = np.gradient(Psi, axis = 0)

#print(Psi)

# plot absolute flow speed
#v = np.sqrt(v_x**2 + v_y**2)
#
#print('Maximum Flow Speed: ', np.max(v))
#
## Normalization:
#v_x = v_x/np.max(v)
#v_y = v_y/np.max(v)

plt.pcolormesh(x, y, Psi.T)
plt.colorbar()
```

```python
v = np.sqrt(vx**2 + vy**2)

plt.pcolormesh(x,y,v)
plt.colorbar()
plt.quiver(x[::3], y[::3], vx[::3,::3].T, vy[::3,::3].T, color = "lightblue")
```

## Gaussian Temperature anomaly

```python
# width of anomaly
sigma = L
DT0 = 3 # [K]

def Gaussian(x,y,sigma):
    
    f = np.exp(-((x**2. + y**2.)/sigma**2.))/np.sqrt(2*np.pi*sigma**2.)
    
    return f

# temperature distribution
DeltaT = DT0*Gaussian(x_mesh,y_mesh, sigma)/np.max(Gaussian(x_mesh,y_mesh, sigma))
DeltaT = (DeltaT-0.5)*2

plt.pcolormesh(x,y,DeltaT, cmap = "plasma")
plt.colorbar()
```

```python
# Put the origin bottom left
x_mesh += -np.min(x)
y_mesh += -np.min(y)

x += -np.min(x)
y += -np.min(y)
```

## Save Model as h5-file - for Python Code

```python
# Create an HDF5 file in write mode ('w')
with h5py.File(f'LibrationBaseFlow.h5', 'w') as f:
    f.create_dataset("x", data = x)
    f.create_dataset("y", data = y)
    f.create_dataset('DeltaT', data = DeltaT)
    f.create_dataset('vx', data = vx[:,:])
    f.create_dataset('vy', data = vy[:,:])
    f.create_dataset('DeltaT0', data = DT0)
    f.create_dataset('v0', data = np.max(np.abs(np.sqrt(vx[:,:]**2 + vy[:,:]**2))))
    f.create_dataset('L', data = L)
```
