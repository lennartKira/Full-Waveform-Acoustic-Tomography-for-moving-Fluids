```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import os

plt.rcParams.update({"font.size": 12})
```

```python
# Initialize variables
nt = int(0)
nr = int(0)
nz = int(0)

# Open the file by Andre and read in nt, nr, nz
# 'Giesecke_Precession0' or 'Giesecke_Precession_TimeAvg'
with open('Giesecke_Precession0.dat', 'r') as file:
    
    nt, nr, nz = [int(val) for val in file.readline().split()]

    # Initialize arrays
    ur = np.zeros((nz, nr), dtype=float)
    uz = np.zeros((nz, nr), dtype=float)
    rr = np.zeros(nr, dtype=float)
    zz = np.zeros(nz, dtype=float)
    tt = np.zeros(nt, dtype=float)

    # Temporary array
    tmp = np.zeros(nr + 1, dtype=float)

    # Read tt(0) and rr array
    tmp = np.array([float(val) for val in file.readline().split()])
    tt[0] = tmp[0]
    for ir in range(nr):
        rr[ir] = tmp[ir + 1]

    # Read zz and uz arrays
    for iz in range(nz):
        tmp = np.array([float(val) for val in file.readline().split()])
        zz[iz] = tmp[0]
        for ir in range(nr):
            uz[iz, ir] = tmp[ir + 1]

    # Read ur array
    for iz in range(nz):
        tmp = np.array([float(val) for val in file.readline().split()])
        for ir in range(nr):
            ur[iz, ir] = tmp[ir + 1]
```

```python
# Obtain Correct Scaling
Re = 1e4 # Reynolds Number (Giesecke)
Po = 0.1 # Poincare Number (Giesecke)

R = 0.5 #[m] Cyl. Radius
nu = 1.5e-5 #[m^2/s] Kin. Viscosity

Omega = nu*Re/R**2 #[1/s] Rot. Frequency 
 
v0 = R*Omega
print(f"Typical Velocity: {v0} m/s")

Re_calc = R**2*Omega/nu
```

```python
from scipy.interpolate import griddata
# rr and zz define the coordinates on the irregular grid
# ur and uz are the data arrays on the irregular grid

# Flatten the rr, zz, ur, and uz arrays to create coordinate pairs
rr_grid, zz_grid = np.meshgrid(rr, zz)
points = np.array([rr_grid.flatten(), zz_grid.flatten()]).T  # Coordinates for each point on the irregular grid
ur_values = ur.flatten() 
uz_values = uz.flatten()

# Define the regular grid
nx = 256
dh = (rr.max() - rr.min())/(nx-1)
x = np.arange(0,nx)*dh + rr.min()  # Regular radial points
ny = int((zz.max()-zz.min())/dh+1) # keep dx = dy = dh and round up for ny
y = np.arange(0,ny)*dh + zz.min()


x_mesh, y_mesh = np.meshgrid(x, y)

# Interpolate `ur` and `uz` onto the regular grid
vx = griddata(points, ur_values, (x_mesh, y_mesh), method='linear').T
vy = griddata(points, uz_values, (x_mesh, y_mesh), method='linear').T

# Put the origin bottom left
x_mesh += -np.min(x)
y_mesh += -np.min(y)

x += -np.min(x)
y += -np.min(y)

# Rescale coordinates such that Radius is R
scale = 2*R/x.max()
x *= scale
y *= scale
x_mesh *= scale
y_mesh *= scale

v0 = 0.1 # Omega*r

vx *= v0
vy *= v0
```

```python
v = np.sqrt(vx**2 + vy**2)
print(f"Typical Flow Speed %.2f m/s" %v0)
L =  (x.max() - x.min())/2 # [m]
print(f"Typical Length Scale %.2f m" %L)

v_inc = 12

plt.figure(figsize = (4.5,3.5))
plt.pcolormesh(x,y,v.T)
plt.colorbar(label = "$u$ [m/s]")
plt.quiver(x[::v_inc],y[::v_inc], vx[::v_inc,::v_inc].T, vy[::v_inc,::v_inc].T, color = 'snow', width = 0.004)
plt.xlabel("$x$ [m]")
plt.ylabel("$y$ [m]")
plt.gca().set_aspect("equal")

#plt.savefig("../Figures/Precession_Model.jpg", dpi = 300, bbox_inches = "tight")
plt.show()
```

## Zero Temperature Anomaly

```python
DeltaT = np.zeros(vx.shape)
DeltaT0 = 0 # [K] - irrelevant for isothermal
```

## Save Model as h5-file - for Python Code

```python
# Create an HDF5 file in write mode ('w')
with h5py.File(f'Precession.h5', 'w') as f:
    f.create_dataset("x", data = x)
    f.create_dataset("y", data = y)
    f.create_dataset('DeltaT', data = DeltaT)
    f.create_dataset('vx', data = vx[:,:])
    f.create_dataset('vy', data = vy[:,:])
    f.create_dataset('DeltaT0', data = DeltaT0)
    f.create_dataset('v0', data = v0)
    f.create_dataset('L', data = L)
```
